## Beispiel - Zwei Würfel

Es wird mit zwei Würfeln gleichzeitig gewürfelt und die Augenzahl addiert.

Erstellen eines Histograms mit der Funktion [plt.hist()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)

````python
N = 10000
wuerfel1 = np.random.randint(1,7,size=N)
wuerfel2 = np.random.randint(1,7,size=N)

wuerfel_summe = wuerfel1+wuerfel2

plt.hist(wuerfel_summe,bins=np.arange(1.5,13.5,1))
````

## Beispiel Normalverteilung

Mittelwert $\mu=0$ und Standardabweichung $\sigma=1$

````python
N = 10000
x = np.random.randn(N)
x.mean(), x.std()
````

#### Erstellen eines Histogramms:

[`hist()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html) - Berechnen und Zeichen eines Histogramms

````python
plt.hist(x,bins=100)
````

Normalverteilung mit speziell vorgegeben Werten für Mittelwert und Standardabweichung können mit der Funktion
[random.normal()](https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html) erzeugt werden, siehe 
[Create an array with a pre determined mean and standard deviation](https://stackoverflow.com/questions/50177594/create-an-array-with-a-pre-determined-mean-and-standard-deviation)

````python
N = 10000
my_mean = 0.5
my_std = 3
x = np.random.normal(loc=my_mean, scale=my_std, size=1000)
x.mean(), x.std()
````

alternative Transformation der von `random.randn()` erzeugten Wert:

````python
x = my_std * np.random.randn(N) + my_mean
x.mean(), x.std()
````

#### Erstellen eines Boxplots

[`boxplot`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html)

````python
plt.boxplot(x,vert=False)
plt.grid(True)
````
