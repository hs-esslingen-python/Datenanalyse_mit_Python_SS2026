# Matplotlib - Beispiel - $\LaTeX$-Beschriftung (sinc-Funktion)


Beispiel: Plotten der Spaltfunktion $\operatorname{sinc}(x)$

Die Spaltfunktion $\operatorname{sinc}(x)$, Details siehe [Sinc-Funktion Wikipedia](https://de.m.wikipedia.org/wiki/Sinc-Funktion) ist definiert als der folgende Quotient:

$$
    y(x) = \operatorname{sinc}(x) = \frac{\sin(x)}{x}
$$

### Python Pakete laden

````python
import numpy as np
import matplotlib.pyplot as plt
````

### Werte x und y für den Plot erzeugen
x-Werte: Vektor mit 1001 Werten im Bereich von $-4\pi$ bis $4\pi$ erzeugen

````python
x = np.linspace(-4*np.pi,4*np.pi, 1001)
#x
````

y-Werte: $\operatorname{sinc}(x)$ - Funktion für die x-Werte berechnen

Warning: Es wird durch Null geteilt !!!

````python
y = np.sin(x)/x
#y
````

### Plotten der Funktion

Hinweis: In Visual Studio Code innerhalb Jupyter Notebook und in Python Skripten werden Plots sichtbar gemacht durch `plt.show()`

````python
plt.figure()         # Erstellen einer neuen Figur
plt.plot(x,y)
````

### Beschriftung der Achsen
#### Beschriftung der x-Achse
Beschriftung der x-Achse mit $\LaTeX$-Labels  an den Stellen $-4\pi$,  $-2\pi$, $0$, $2\pi$ und $4\pi$. -> `plt.xticks()`

Bei der Definition der Beschriftung werden raw-strings r"..." (Buchstabe 'r' wird der Zeichenkette vorangestellt) verwendet, um folgende Warnungen zu vermeiden:
```Text

```

````python
plt.figure()         # Erstellen einer neuen Figur 
plt.xticks([-4*np.pi,-2*np.pi,0,2*np.pi,4*np.pi],[r'$-4\pi$',r'$-2\pi$',r'$0$',r'$2\pi$',r'$4\pi$'])
plt.plot(x,y)
````

#### Beschriftung der y-Achse

Beschriftung der y-Achse mit $\LaTeX$-Labels nur an den Stellen $0$ und $1$ mit `plt.yticks()`.
<br>
Mit der Befehl  `plt.yticks([0,1],['$0$','$1$'])` wird die y-Achse nur an den Stelle 0 und 1 $\LaTeX$-Labels beschriftet.

````python
plt.figure()     # Erstellen einer neuen Figur 
plt.xticks([-4*np.pi,-2*np.pi,0,2*np.pi,4*np.pi],[r'$-4\pi$',r'$-2\pi$',r'$0$',r'$2\pi$',r'$4\pi$'])
plt.yticks([0,1],['$0$','$1$'])
plt.plot(x,y)
````

### Größe der Figur und der Achsenbeschriftung ändern

+ Größe der Figur `plt.figure(figsize=(11,4))`
<br>
Mit dem Parameter `figsize` in Matplotlib wird die Größe einer Grafik (Figure) festgelegen. 
Dieser Parameter wird als Tupel (Breite, Höhe) in Zoll (inches) an die Funktion plt.figure() übergeben. 
Das Standardmaß ist (6.4, 4.8) Zoll
+ Größe der Achsenbeschriftung `plt.xticks(...,fontsize=17)` und `plt.xticks(...,fontsize=17)`
<br>
Mit plt.xticks(..., fontsize=17) wird in Matplotlib die Schriftgröße der x-Achsen-Beschriftungen (Tick-Labels) direkt festgelegen.

<!--
Matplotlib's runtime configuration (rc) options - Matplotlib rcParams

- breiterer Plot:  `plt.rcParams['figure.figsize'] = (11,4)`

- Schriftgröße erhöhen: `plt.rcParams['font.size'] = 17`

siehe auch [Customizing Matplotlib with style sheets and rcParams](https://matplotlib.org/stable/tutorials/introductory/customizing.html)

[matplotlib.rcParams](https://matplotlib.org/stable/api/matplotlib_configuration_api.html#matplotlib.rcParams) ist ein Python Dictionary und enthält die aktuell verwendeten Einstellparameter für Matplotlib.
-->

````python

# Erstellen einer neuen Figur und Festlegen der Größe
plt.figure(figsize=(11,4))


# Beschriftung der x-Achse und der y-Achse und Verändern der Schriftgröße
plt.xticks([-4*np.pi,-2*np.pi,0,2*np.pi,4*np.pi],[r'$-4\pi$',r'$-2\pi$',r'$0$',r'$2\pi$',r'$4\pi$'],fontsize=17)
plt.yticks([0,1],['$0$','$1$'],fontsize=17)


# Plotten x,y Werte
plt.plot(x,y)

````

### Legende
- `plt.plot(x,y, label=r'$\frac{\sin(x)}{x}$')`  - für $\LaTeX$ Beschriftung einen raw String r"..." verwenden.
- `plt.legend(loc='best', fontsize=24)` - bestmögliche Position ohne Überdecken der geploteten Linie; spezifische Schriftgröße vorgehen

````python
# Erstellen einer neuen Figur und Festlegen der Größe
plt.figure(figsize=(11,4))

# Beschriftung der x-Achse und der y-Achse und Verändern der Schriftgröße
plt.xticks([-4*np.pi,-2*np.pi,0,2*np.pi,4*np.pi],[r'$-4\pi$',r'$-2\pi$',r'$0$',r'$2\pi$',r'$4\pi$'],fontsize=17)
plt.yticks([0,1],['$0$','$1$'],fontsize=17)


# Plotten x,y Werte
plt.plot(x,y, label=r'$\frac{\sin(x)}{x}$') 

# Legende
plt.legend(loc='best', fontsize=24)
                          
````

### Titel 

`plt.title(r'$=\frac{\sin(x)}{x}$ - Funktion')`

````python
# Plotgröße
plt.figure(figsize=(11,4))

# Beschriftung der x-Achse und der y-Achse und Verändern der Schriftgröße
plt.xticks([-4*np.pi,-2*np.pi,0,2*np.pi,4*np.pi],[r'$-4\pi$',r'$-2\pi$',r'$0$',r'$2\pi$',r'$4\pi$'],fontsize=17)
plt.yticks([0,1],['$0$','$1$'],fontsize=17)


# Plotten x,y Werte
plt.plot(x,y, label=r'$\frac{\sin(x)}{x}$') 

# Title
plt.title(r'$=\frac{\sin(x)}{x}$ - Funktion')

# Legende
plt.legend(loc='best', fontsize=24);

# Gitter 
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

````

### horizontale Linie bei y=0:

`plt.axhline(0,color='black',lw=2)`

````python
# Plotgröße
plt.figure(figsize=(11,4))


# Beschriftung der x-Achse und der y-Achse und Verändern der Schriftgröße
plt.xticks([-4*np.pi,-2*np.pi,0,2*np.pi,4*np.pi],[r'$-4\pi$',r'$-2\pi$',r'$0$',r'$2\pi$',r'$4\pi$'],fontsize=17)
plt.yticks([0,1],['$0$','$1$'],fontsize=17)


# Plotten x,y Werte
plt.plot(x,y, label=r'$\frac{\sin(x)}{x}$',lw=2) 

# schwarze horizontale Linie bei y=0 mit einer Linienstärke von 2
plt.axhline(0,color='black',lw=2)

# Title
plt.title(r'$=\frac{\sin(x)}{x}$ - Funktion')

plt.tight_layout()


# Gitter 
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Legende
plt.legend(loc='best', fontsize=24)

````

### Abspeichern als png-Datei

`plt.savefig('sinc-function.png')`

Die Parameter `facecolor='white', transparent=False` behebt das Problem, dass das erzeugte png-Bild einen schwarzen Hintergrund hat.

````python
# Plotgröße
plt.figure(figsize=(11,4))

# Beschriftung der x-Achse und der y-Achse und Verändern der Schriftgröße
plt.xticks([-4*np.pi,-2*np.pi,0,2*np.pi,4*np.pi],[r'$-4\pi$',r'$-2\pi$',r'$0$',r'$2\pi$',r'$4\pi$'],fontsize=17)
plt.yticks([0,1],['$0$','$1$'],fontsize=17)


# Plotten x,y Werte
plt.plot(x,y, label=r'$\frac{\sin(x)}{x}$',lw=2) 

# schwarze horizontale Linie bei y=0 mit einer Linienstärke von 2
plt.axhline(0,color='black',lw=2)

# Title
plt.title(r'$=\frac{\sin(x)}{x}$ - Funktion')

plt.tight_layout()

# Legende
plt.legend(loc='best', fontsize=24);

# Abspeichern als png-Datei
plt.savefig('sinc-function.png',facecolor='white', transparent=False)
````

### Objekt-orientiertes Interface  (OO-Interface)

````python
# Plotgröße
fig, ax = plt.subplots(figsize=(11,4))

# Beschriftung der x-Achse und der y-Achse 
ax.set_xticks([-4*np.pi,-2*np.pi,0,2*np.pi,4*np.pi],[r'$-4\pi$',r'$-2\pi$',r'$0$',r'$2\pi$',r'$4\pi$'])
ax.set_yticks([0,1],['$0$','$1$'])

# Verändern der Schriftgröße über das Achsenobjekt mit tick_params
ax.tick_params(axis='x', labelsize=17)
ax.tick_params(axis='y', labelsize=17)

# Plotten der x,y Werte
ax.plot(x,y, label=r'$\frac{\sin(x)}{x}$',lw=2) 

# schwarze horizontale Linie bei y=0 mit einer Linienstärke von 2
ax.axhline(0,color='black',lw=2)

# Title
ax.set_title(r'$=\frac{\sin(x)}{x}$ - Funktion')

fig.tight_layout()

# Legende
ax.legend(loc='best', fontsize=24);
````
