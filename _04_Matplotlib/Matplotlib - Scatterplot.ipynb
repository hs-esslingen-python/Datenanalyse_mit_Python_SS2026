```python
import numpy as np
import matplotlib.pyplot as plt
```

## Beispiel Daten erzeugen

4 Vektoren (Listen) mit jeweils 7 Werten als _Numpy Arrays_ erzeugen:

```python
x = np.array([ 1.0,  4.3,  5.2,   2.1,  7.5,  6.9,  9.9])
y = np.array([ 2.3,  3.3,  6.3,   5.8,  4.3,  3.3,  8.3])
w = np.array([10.1, 30.2,  50.4, 70.2, 90.5, 60.0, 20.0])
z = np.array([ 1.1,  5.4,   6.5,  7.2,  8.5,  9.5,  2.0])
```

## Scatterplot erstellen

Mit der Funktion [`plt.scatter(x,y)`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html) wird ein Scatterplot erstellt, wobei die x und y-Koordinaten der Punkte in den Parametern `x` und `y` übergeben werden:

```python
plt.scatter(x,y)
```

### Verändern der Markergröße
Größe der Punkte (Marker) wird durch den Parameter `s=` vorgegeben:

```python
plt.scatter(x,y,s = 100)
```

Jeder Punkt kann einzeln in der Größe angepasst werden

```python
plt.scatter(x,y,s = w)
```

## Verändern der Markerfarbe

Die Farbe jedes Punktes kann durch den Parameter `c=` in Verbindung mit einer Colormap `cmap=` verändert werden

[Matplotlib-colormaps](https://matplotlib.org/stable/tutorials/colors/colormaps.html)

```python
plt.scatter(x,y,s=w*20, c=z,cmap='hsv')
```

weitere Mögliche Anpassungen:
- Punkte mit einem schwarzen Rand versehen: `edgecolor='black'` und Strichstärke `linewidth=1.5`
- Farbkonstrast reduzieren durch erhöhen der Durchlässigkeit (Transparenz) `alpha=`
- Bereitstellen einer Legende mit Farbverlauf: `plt.colorbar`

```python
plt.scatter(x,y,s=w*20, c=z,cmap='coolwarm',edgecolor='black', linewidth=1.5,alpha=0.6)
#plt.scatter(x,y,s=w*20, c=z,cmap='hsv',edgecolor='black', linewidth=1.5,alpha=0.6)
cbar = plt.colorbar()
cbar.set_label('Meine Farbpalette')
```

Damit ist eine Darstellung von vier Eigenschaften in einem Plot möglich:
- x Position
- y Position
- Größe des Punktes
- Farbe des Punktes

### Beispiel - Visualisieren des Verhaltens einer unabhängigen und einer abhängigen Variablen 

- x - unabhängige Variable
- y - abhängige Variable

$
  y(x) = 2x + 3x^2 + 0.5x^3 + noise
$

```python
# Beispieldaten generieren
N = 100
np.random.seed(7)
x = np.random.randn(N)
y = 2*x + 3*x**2 + 0.5*x**3 + 2*np.random.randn(N)
```

```python
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```

```python
plt.plot(x)
plt.plot(y)
```
