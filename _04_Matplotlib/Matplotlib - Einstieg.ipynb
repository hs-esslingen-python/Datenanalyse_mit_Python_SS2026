## Aufbau eines Matplotlib Graphen
Ein Matplotlib Graph besteht aus einer 
- "Leinwand" [Figure](https://matplotlib.org/stable/api/_as_gen/matplotlib.figure.Figure.html#matplotlib.figure.Figure) in der 
- eine oder mehrere "Diagramme mit Achsen" [Axes](https://matplotlib.org/stable/api/axes_api.html#matplotlib.axes.Axes) plaziert werden.

Man beachte:
- Axis - die Achse (Singular)
- Axes - die Achsen (Plural)

## Minimal Beispiel zur Einführung

````python
fig, ax = plt.subplots()             # Erzeugen einer Figure (fig) mit einem Diagramm (ax) 
x = [1, 2, 3, 4]                     # Werte für die x-Achse (Python Liste)              
y = [1, 4, 2, 3]                     # Werte für die y-Achse (Python Liste)
ax.plot(x,y)                         # Plot der Daten als Kurve y über x im Diagramm (ax)
#plt.show()                           # Anzeigen des Diagramms in einem Fenster (nicht notwendig in Jupyter Notebooks)
````

Anmerkung: In einem Python Skript muss zum Darstellen des Plots der folgende Befehl ausgeführt werden, ansonsten wird der Plot nicht angezeigt:
```Python
plt.show()
```

### Minimal Beispiel - Es geht auch einfacher - nur in einer Zeile:

````python
plt.plot(x,y)
````

## Aufbau einer Figure 

Das folgende Bild veranschaulicht die Elemente einer Figure mit einem Diagramm.

<img src="https://matplotlib.org/stable/_images/anatomy.png" width="600" >


+ Figure -> `fig = plt.figure()` -> Zeichenblatt
    + Axes -> `ax = fig.subplots()` -> Diagramm
        + Line -> `ax.plot()`  -> Linien-Plot
        + Markers -> `ax.scatter()` -> Punkte-Plot
        + Grid  -> `ax.grid()`  -> Gitternetz
        + Title -> `ax.set_title()` -> Diagrammtitel
        + xlabel, ylabel -> `ax.set_xlabel()`, `ax.set_ylabel()` -> Achsbeschriftung
        + Legend -> `ax.legend()` -> Beschreibung zu den Linien und Punkten
        + Spine -> `ax.spines` -> Steuerung des Diagramm-Rahmens
        + x Axis, y Axis -> `ax.xaxis` und `ax.yaxis` -> Steuerung der x,y-Achsen
            + Major tick -> `ax.xaxis.set_major_locator()`
            + Minor tick -> `ax.xaxis.set_minor_locator()`
            + Major tick label -> `ax.xaxis.set_major_formatter()`
            + Minor tick label -> `ax.xaxis.set_minor_formatter()`


Lineplot

````python
x = [1, 2, 3, 4]                # Werte für die x-Achse (Python Liste)              
y = [1, 4, 2, 3]                # Werte für die y-Achse (Python Liste)


fig = plt.figure()              # Erzeugen einer neuen Figure (fig)
ax = fig.subplots()             # Erzeugen eines Diagramms (ax) in der Figure (fig) 
ax.plot(x,y)                    # Plot der Daten als Kurve y über x im Diagramm (ax)
ax.legend(["Daten"])            # Hinzufügen einer Legende zum Diagramm (ax) mit der Bezeichnung "Daten"
ax.grid()                       # Hinzufügen eines Gitters zum Diagramm (ax)
ax.set_title("Line-Plot")       # Setzen des Titels des Diagramms (ax) auf "Line-Plot"
ax.set_xlabel("x-Achse")        # Setzen der Beschriftung der x-Achse des Diagramms (ax) auf "x-Achse"
ax.set_ylabel("y-Achse")        # Setzen der Beschriftung der y-Achse des Diagramms (ax) auf "y-Achse"
````

Punkte-Plot (Scatter-Plot)

````python
x = [1, 2, 3, 4]                # Werte für die x-Achse (Python Liste)              
y = [1, 4, 2, 3]                # Werte für die y-Achse (Python Liste)


fig = plt.figure()              # Erzeugen einer neuen Figure (fig)
ax = fig.subplots()             # Erzeugen eines Diagramms (ax) in der Figure (fig) 
ax.scatter(x,y)                 # Plot der Daten als Punkte y über x im Diagramm (ax)
ax.legend(["Daten"])            # Hinzufügen einer Legende zum Diagramm (ax) mit der Bezeichnung "Daten"
ax.grid()                       # Hinzufügen eines Gitters zum Diagramm (ax)
ax.set_title("Scatter-Plot")    # Setzen des Titels des Diagramms (ax) auf "Scatter-Plot"
ax.set_xlabel("x-Achse")        # Setzen der Beschriftung der x-Achse des Diagramms (ax) auf "x-Achse"
ax.set_ylabel("y-Achse")        # Setzen der Beschriftung der y-Achse des Diagramms (ax) auf "y-Achse" 
````

- eine oder mehrere Diagramme (axes)

### Es gibt zwei unterschiedliche Progammiermöglichkeiten zum Erstellen von Plots mit Matplotlib:
- objekt orientiertes Interface  (OO-Interface)
- funktionales Interface  (pyplot interface)

### Objekt-orientiertes Interface  (OO-Interface)

bietet mehr Flexibilität bei der Gestaltung; bedeutet aber mehr Schreibaufwand

````python
x = np.linspace(0, 2, 100)           # x-Werte anlegen

fig, ax = plt.subplots()             # figure und axes-Objekt anlegen.
ax.plot(x, x,    label='linear')     # erste  Kurve in das axes-Objekt plotten
ax.plot(x, x**2, label='quadratic')  # zweite Kurve in das axes-Objekt plotten
ax.plot(x, x**3, label='cubic')      # dritte Kurve in das axes-Objekt plotten
ax.set_xlabel('x label')             # Beschriftung für die X-Achse des axes-Objektes
ax.set_ylabel('y label')             # Beschriftung für die Y-Achse des axes-Objektes
ax.set_title("Simple Plot")          # Titel dem axes-Objektes geben
ax.legend()                          # Anlegen der Legende Add a legend.
ax.grid(True)                        # Gitternetzlinien
````

### Funktionales Interface (pyplot-style)

weniger Schreibaufwand, möglicherweise Einschränkungen bei der Gestaltung

````python
x = np.linspace(0, 2, 100)


plt.plot(x, x,    label='linear')  
plt.plot(x, x**2, label='quadratic')  
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
plt.grid(True)
````

#### Beispielplot mit "Format-String"

Mit dem Format-String `'o-r'` wird eine rote durch gezogene Linie mit Kreisen an den Messpunkten dargestellt:
```Python
plt.plot(x, x,   'o-r',   label='linear')     
```
 
Mit dem Format-String `'^--b'` wird eine blaue gestrichelte Linie mit nach oben stehenden Dreiecken an den Messpunkten dargestellt:
```Python
plt.plot(x, x**2, '^--b', label='quadratic')  
```

Mit dem Format-String `'s:g'` wird eine grüne gepunktete Linie mit Quadraten an den Messpunkten dargestellt:
```Python
plt.plot(x, x**3, 's:g',  label='cubic')
```

````python
x = np.linspace(0, 2, 11)

plt.plot(x, x,   'o-r',   label='linear')     # rote durch gezogene Linie mit Kreisen an den Messpunkten
plt.plot(x, x**2, '^--b', label='quadratic')  # blaue gestrichelte Linie mit nach oben stehenden Dreiecken an den Messpunkten
plt.plot(x, x**3, 's:g',  label='cubic')      # grüne gepunktete Linie mit Quadraten an den Messpunkten 


plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
plt.grid(True)
````

### Gestaltung der Kurven durch explizite Parameter

- `marker=`  - Angabe des Markers
- `linestyle=` - Angabe des Linienmusters  (kann auch `ls=` abkürzt werden)
- `color=` - Angabe der Farbe der Linie

Beispielplot mit Auswahl wie oben nur mit Angabe durch explizite Parameter:

````python
x = np.linspace(0, 2, 11)

plt.plot(x, x,    marker='o', linestyle='-',  color='r',  label='linear')     
plt.plot(x, x**2, marker='^', linestyle='--', color='b',  label='quadratic')  
plt.plot(x, x**3, marker='s', ls=':',         color='g',  label='cubic')


plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
plt.grid(True)
````

### Größe der Figure verändern

Mit dem Parameter `figsize=` beim Aufruf von `plt.figure()` kann die Größe der Figure vorgegeben werden. Übergeben wird eine Tupel mit (Breite, Höhe) in inch (1 inch sind 2,54cm).

```Python 
   plt.figure(figsize=(Breite,Höhe))
```

### Weitere Parameter von Plot():
- `linewidth=` oder `lw=` - Liniendicke
- `alpha=` - Transparenz [0..1]  0=nicht sichtbar; 1=voll sichtbar
- `linestyle=` oder `ls=` - Linienmuster

Marker
- `marker=`  - Form
- `markersize=` - Größe 
- `markerfacecolor=` - Innenfarbe
- `markeredgewidth=` - Randgröße
- `markeredgecolor=` - Randfarbe

Darstellungsbereich einschränken:
- `plt.xlim(untere Grenze, obere Grenze)` - x-Achse
- `plt.ylim(untere Grenze, obere Grenze)` - y-Achse



#### Beispielplot

````python
x = np.linspace(0, 2, 11)

plt.figure(figsize=(12,8))  # Breite: 12 inch und Höhe: 8 inch

plt.plot(x, x,  linestyle='-',  linewidth = 5, color='orange',  
                marker='o', markersize=15, markerfacecolor='blue', markeredgewidth=4, markeredgecolor='red',        
                label='linear')     
plt.plot(x, x**2, linestyle='--', linewidth = 5, color=(0.54,0,1.0),  
                  marker='^', markersize=10, markerfacecolor='yellow', markeredgecolor='black',   
                  label='quadratic',alpha=.4)  
plt.plot(x, x**3, marker='s', markersize=10, linestyle=':',  linewidth = 5,  color='#bfff00',  label='cubic',alpha=.3)

plt.xlim(0.5,1.5)  # Einschränken der x-Achse
plt.ylim(0,4)      # Einschränken der y-Achse

plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend(loc='upper center')
plt.grid(True)
````

## Mehrere Plot-Diagramme zeichnen

Mit der Funktion `plt.subplot(Anzahl Zeilen, Anzahl Spaltem, Index)` (Achtung _**subplot**_ in der Einzahl!!!) lassen sich in einer Figure mehrer Plots orientiert an einer Matrix (Zeilen, Spalten) einzeichnen.
Solange die Zahlen einstellig sind, können sie als eine dreistellige Zahl geschrieben werden, z.B. `plt.subplot(3,1,1)` kann auch als `plt.subplot(311)` geschrieben werden.

[Matplotlib-subplot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html)

[Matplotlib-tight_layout](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.tight_layout.html)


#### Beispielplot mit 3 vertikal angeordneten Subplots 

````python
x = np.linspace(0, 2, 11)

plt.figure(figsize=(10,8))

# 1. Subplot
plt.subplot(3,1,1) # 3 Zeilen und 1 Spalte, Subplots Nr. 1 
plt.plot(x, x,    marker='o', linestyle='-',  color='orange',  label='linear') 
plt.ylabel('y label')
plt.legend()
plt.grid(True)
plt.title("Simple Plot")  # der oberste Subplot bekommt eine Überschrift

# 2. Subplot
plt.subplot(3,1,2) # 3 Zeilen und 1 Spalte, Subplots Nr. 2
plt.plot(x, x**2, marker='^', linestyle='--', color=(0.54,0,1.0) ,  label='quadratic')  
plt.ylabel('y label')
plt.legend()
plt.grid(True)

# 3. Subplot
plt.subplot(3,1,3) # 3 Zeilen und 1 Spalte, Subplots Nr. 3 
plt.plot(x, x**3, marker='s', linestyle=':',  color='#bfff00',  label='cubic')
plt.title("Unterer Plot")
plt.xlabel('x label')
plt.ylabel('y label')
plt.legend()
plt.grid(True)

plt.tight_layout() # Überlappung möglichst vermeiden 
````

### Jetzt das gleiche mit dem Objekt-orientierten Interface  (OO-Interface)

Mit der Top-Level-Funktion `plt.subplots`   wird ein Figure-Objekt und mehrere angeordnete axes-Objekte erstellen.
(**subplots** steht hier jetzt in der Mehrzahl !!!)

[Matplotlib-subplots](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html)

Parameter:
- `nrows=` Anzahl der Zeilen 
- `ncols=` Anzahl der Spalten
- `figsize=` Größe des Figure-Objekts
- `sharex=True` bzw. `sharey=True` Skalierung auf den x- bzw. y-Achsen synchronisieren 

````python
fig, ax = plt.subplots()       # eine Figure mit einem Diagramm (Axes)
ax
````

````python
fig, axs = plt.subplots(2, 2)  # eine Figure mit 4 Diagrammen angeordnet in 2 Spalten und 2 Zeilen
axs
````

````python
fig, axs = plt.subplots(nrows = 3, ncols=3,figsize=(14,8))
axs
````

Was ist `axs`?

````python
axs
````

`axs` ist Array von Arrays mit einzelen `AxesSubplot`-Objekten.

Diese werden wir durch "Tuple-unpacking" gleich auf mehrere Variablen aufteilen, so bekommen wir einen anschaulichen Zugriff auf die einzelnn `AxesSubplot`-Objekte.

````python
((ax1,ax2,ax3),(ax4,ax5,ax6),(ax7,ax8,ax9)) = axs
ax1
````

````python
fig, (ax1,ax2,ax3) = plt.subplots(nrows = 3, ncols=1,figsize=(14,8))
ax1
````

#### Beispielplot - Mehrere Subplots erstellt mit `plt.subplots()`

Man beachte die Mehrzahl bei `plt.subplots()`

````python
x = np.linspace(0, 2, 11)

fig, (ax1,ax2,ax3) = plt.subplots(nrows = 3, ncols=1,figsize=(12,8),sharex=True)

fig.tight_layout() # Überlappung von Diagrammen möglichst vermeiden 


# ax1 - linear - Subplot oben
ax1.plot(x, x,    marker='o', linestyle='-',  color='orange',  label='linear') 
ax1.set_title("Simple Plot")
ax1.set_ylabel('y label')
ax1.legend()
ax1.grid(True)

# ax2 - quadratisch - Subplot Mitte
ax2.plot(x, x**2, marker='^', linestyle='--', color=(0.54,0,1.0) ,  label='quadratic')  
ax2.set_ylabel('y label')
ax2.legend()
ax2.grid(True)

# ax3 - kubisch - Subplot unten
ax3.plot(x, x**3, marker='s', linestyle=':',  color='#bfff00',  label='cubic')
ax3.set_xlabel('x label')
ax3.set_ylabel('y label')
ax3.legend()
ax3.grid(True)
````

Jetzt schränken wir den Bereich einer x-Achse mit der Anweisung `ax3.set_xlim(0.5,1.5)` ein.

Durch den Parameter `sharex=True` werden alle x-Achsen gleichzeitig eingeschränkt. 

#### Beispielplot - alle x-Achsen gleichtmäßig mit Parameter `sharex=True` im Bereich einschränken

````python
x = np.linspace(0, 2, 11)

fig, (ax1,ax2,ax3) = plt.subplots(nrows = 3, ncols=1,figsize=(14,8),sharex=True) # alle x-Achsen gleichmäßig einschränken

fig.tight_layout() # Überlappung möglichst vermeiden 

# ax1
ax1.plot(x, x,    marker='o', linestyle='-',  color='orange',  label='linear') 
ax1.set_title("Simple Plot")
ax1.set_ylabel('y label')
ax1.legend()
ax1.grid(True)

# ax2
ax2.plot(x, x**2, marker='^', linestyle='--', color=(0.54,0,1.0) ,  label='quadratic')  
ax2.set_ylabel('y label')
ax2.legend()
ax2.grid(True)

# ax3
ax3.plot(x, x**3, marker='s', linestyle=':',  color='#bfff00',  label='cubic')
ax3.set_xlabel('x label')
ax3.set_ylabel('y label')
ax3.legend()
ax3.grid(True)

ax3.set_xlim(0.5,1.5)
````

### Abspeichern der Figure als Bild in einer Datei

Mit der `.savefig()`-Methode des Figure-Objektes wird der Inhalt als Bild in eine Datei geschrieben.

Mit dem Parameter `dpi = ` wird die Auflösung von pixels per inch vorgegeben.

[Matplotlib-savefig](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.savefig.html)

Endung des Dateinames bestimmt das verwendete Bild Format
- .png

### Diagramme beliebig in einer Figure plazieren

Wir erzeugen mit `plt.figure()` eine leere Figure und können dann mit `.add_axes()` ein Diagramm in der Figure plazieren, dessen Position und Größe durch Übergabe einer Liste `[left, bottom, width, height]` als Paramter mit Einträgen definiert wird: 
- left : x coordinate lower left corner
- bottom: y coordinate lower left corner
- width : Breite
- height : Höhe

Koordinatensystem der Figure:
- Der Ursprung (0,0) ist links unten und 
- rechts oben ist der Punkt (1.0,1.0)

#### Beispielplot 

````python
x = np.linspace(0, 2, 11)

fig = plt.figure()             # eine leere Figure; noch ohne Diagramme (Axes)

ax1 = fig.add_axes([0,0,1,1]) # ein Diagramm welches die gesamte Figure ausfüllt
ax1.plot(x, x,    marker='o', linestyle='-',  color='orange',  label='linear') 

ax2 = fig.add_axes([0.1,0.6,0.5,0.3]) # eine Diagrammmit Ursprung bei (0.1, 0.6), einer Breite von 0.5 und einer Höhe von 0.3
ax2.plot(x, x**2, marker='^', linestyle='--', color=(0.54,0,1.0) ,  label='quadratic')  

````

### Style ändern

Ein Style setzt eine Reihe von Standardeinstellungen (Farben, Liniendicken, Schriftgrößen, etc.). Mit `plt.style.use()` wird ein Style aktiviert und beeinflusst alle nachfolgenden Plots.

**Wichtig:** Sobald Sie einen Style mit `plt.style.use()` setzen, bleibt dieser für alle folgenden Plots aktiv, bis Sie einen anderen Style setzen oder die Einstellungen zurücksetzen.

````python
# Hilfsfunktion: Plot mit einem bestimmten Style visualisieren
def show_style(style_name, x_data=None, y_data=None):
    """
    Visualisiert einen Style mit einem Beispielplot.
    
    Parameter:
    - style_name: Name des Styles (z.B. 'classic', 'seaborn-v0_8')
    - x_data, y_data: Optionale Daten (verwenden Standard-Daten wenn None)
    """
    # Standard-Daten wenn keine Daten übergeben
    if x_data is None:
        x_data = np.linspace(0, 2, 11)
        y_data = [x_data, x_data**2, x_data**3]
    
    plt.style.use(style_name)
    plt.figure(figsize=(10, 4))
    
    plt.plot(x_data, y_data[0], marker='o', label='linear')
    plt.plot(x_data, y_data[1], marker='^', label='quadratic')
    plt.plot(x_data, y_data[2], marker='s', label='cubic')
    
    plt.title(f"Style: {style_name}")
    plt.xlabel('x label')
    plt.ylabel('y label')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()

# Beispiel: Einen Style aktivieren
show_style('seaborn-v0_8')
````

#### Comic-Style (xkcd)

Matplotlib hat auch einen speziellen Comic-Style, der mit einem Context-Manager verwendet wird:

**Hinweis:** Die Fehlermeldungen zu fehlenden Fonts (`findfont: Font family 'xkcd' not found...`) sind harmlos – sie erscheinen nur, weil spezielle Comic-Fonts nicht installiert sind. Der Plot wird trotzdem korrekt angezeigt, Matplotlib nutzt dann automatisch Fallback-Fonts.


````python
import warnings
import logging

# Alle matplotlib-Warnungen unterdrücken
warnings.filterwarnings('ignore', category=UserWarning)
warnings.filterwarnings('ignore', message='.*findfont.*')

# Matplotlib-Logger auf ERROR setzen, um Meldungen zu unterdrücken
logging.getLogger('matplotlib').setLevel(logging.ERROR)

x = np.linspace(0, 2, 11)

with plt.xkcd():
    plt.figure(figsize=(10, 4))
    plt.plot(x, x, marker='o', label='linear')
    plt.plot(x, x**2, marker='^', label='quadratic')
    plt.plot(x, x**3, marker='s', label='cubic')
    plt.title('Comic Style (xkcd)')
    plt.xlabel('x label')
    plt.ylabel('y label')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
````

#### Style zurücksetzen

Sobald ein Style aktiviert ist, bleibt dieser für alle nachfolgenden Plots bestehen. Um die vordefinierten Matplotlib-Einstellungen wiederherzustellen:

**Wichtig:** Das Zurücksetzen der rcParams setzt auch die Jupyter-Anzeige zurück. Daher wird `%matplotlib inline` erneut aktiviert, damit Plots wieder automatisch angezeigt werden.

````python
# Matplotlib Standardeinstellungen wiederherstellen
import matplotlib as mpl
from IPython import get_ipython

mpl.rcParams.update(mpl.rcParamsDefault)

# Wichtig: %matplotlib inline erneut aktivieren, damit Plots wieder angezeigt werden
ipython = get_ipython()
if ipython:
    ipython.run_line_magic('matplotlib', 'inline')
````
