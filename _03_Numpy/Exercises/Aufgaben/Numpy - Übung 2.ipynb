````python
import numpy as np
import matplotlib.pyplot as plt
````

## Schwarzweißbild Beispiel


Vorbereitung:
<br>
Es wird zuerst ein schwarzes Bild mit 5 Pixel Höhe und mit 10 Pixel Breite erzeugt: 

-> Dazu wird ein 2D _NumPy Array_ aus Nullen mit erster Dimension von 5 (Höhe) und zweiter Dimension von 10 (Breite) angelegt.

````python
# Numpy Array erzeugen
Hoehe = 5
Breite = 10
img = np.zeros((Hoehe,Breite))

# Bild anzeigen
plt.imshow(img,cmap='gray');
````

Danach werden folgende Pixel gesetzt:
- links oben auf Intensität 1 (weiß)
- links unten auf Intensität 0.75 (helles grau)
- rechts oben auf Intensität 0.5 (dunkles grau)
- rechts unten auf Intensität 0.25 (sehr dunkles grau) 


````python
img = np.zeros((5,10))
img[0,0] = 1.0      # links oben -> weiß
img[-1,0] = 0.75    # links unten -> hellgrau
img[0,-1] = 0.50    # rechts oben -> mittelgrau
img[-1,-1] = 0.25   # rechts unten -> dunkelgrau
plt.imshow(img,cmap='gray');
````

## Farbbild - Beispiel

Es wird zunächst ein schwarzes Bild mit 5 Pixel Höhe und mit 10 Pixel Breite und drei Farbkanälen erzeugt: 

-> Dazu wird ein 3D _NumPy Array_ aus Nullen mit einer
+ ersten Dimension von 5 (Höhe), mit einer
+ zweiten Dimension von 10 (Breite) und mit einer
+ dritten Dimension von 3 (Farbkanäle) angelegt.


````python
img = np.zeros((5,10,3))
plt.imshow(img);
````

Jetzt werden folgende Pixel gesetzt:
- links oben auf rot
- links unten auf grün
- rechts oben auf blau 
- rechts unten auf gelb  (gelb wird aus rot und grün gemischt)

````python
img = np.zeros((5,10,3))
img[0,0,0] = 1.0   # links oben auf rot
img[-1,0,1] = 1.0   # links unten auf grün 
img[0,-1,2] = 1.0   # rechts oben auf blau 
img[-1,-1] = (1.0,1.0,0)   # rechts untern auf gelb (rot auf 1, grün auf 1, blau auf 0) 

plt.imshow(img);
````
