Datentyp Tupel  - "nicht veränderbare Liste"

Tupels (Tuples engl.) werden verwendet, um mehrere Element in einer einzigen Variablen zu speichern, die später nicht mehr verändert werden muss.

Ein Tupel ist im Vergleich zu einer Liste schneller in der Ausführung und benötigt weniger Speicherplatz.

### Datentyp Tupel  - "nicht veränderbare Liste"

Definition durch runde Klammern `()`

```python
tupel_3 = (1,2,3)
tupel_3
```

Datentyp `tuple`

```python
type(tupel_3)
```

Der Zugriff auf ein Element aus einem Tupel erfolgt durch eckige Klammern `[]` und durch Angabe eines Index.
Der Index beginnt mit der Zahl 0.

```python
tupel_3[0]
```

```python
tupel_3[1]
```

```python
tupel_3[2]
```

Zugriff auf einen Index der nicht mehr im Tupel liegen würde, erzeugt eine `IndexError` mit der Fehlermeldung "tuple index out of range" 

```python
tupel_3[3]
```

Versucht man, ein Element eines Tupels zu ändern, reagiert Python mit einem `TypeError` der Fehlermeldung "'tuple' object does not support item assignment" darauf:

```python
tupel_3[1] ='neu'
```

#### `tuple()` - Funktion
Die `tuple()` - Funktion dient zur Erzeugung eines Tupels aus anderen Objekten - hier einer Liste `[1,2,3]`, die wir aber erst später behandeln.

```python
tupel_3b = tuple([1,2,3])
tupel_3b
```

Im Vergleich dazu wird eine Liste durch eckige Klammern dargestellt:

```python
[1,2,3]
```

Mit der Funktion `len()` kann die Anzahl der Elemente im Tupel bestimmen:

```python
len(tupel_3b)
```
