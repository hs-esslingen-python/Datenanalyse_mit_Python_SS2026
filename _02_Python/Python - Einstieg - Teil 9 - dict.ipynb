Schlüssel und Datenwert - Paare (key, value - pairs)

https://realpython.com/python-dicts/

### Definition von Dictionaries 

**Möglichkeit 1:** mit geschweiften Klammern `{}` und dem Doppelpunkt zur Trennung von Schüssel (key) und Werten (value).

```python
dict1 = {'key1':'value1', 'key2':'value2', 'key3':'value3'}
dict1
```

Unterschied zur Definition von _Sets_ klarmachen. Beide verwenden `{}` bei der Definition.
- _Set_: Auflistung der Elemente des Sets durch Kommas getrennt
- _Dictionary_: Auflisten von Schlüssel-Wert-Paaren durch Kommas getrennt. Doppeltpunkt trennt Schlüssel und Wert. 

Der Datentyp eine Python Dictonarys ist `dict`

```python
type(dict1)
```

**Spezialfall:** 
Definition eines Dictionaries, wo der Schlüssel `'key1'` in der Definition mehrfach vorkommt.
<br>
Verhalten: Die Definition wird von links nach rechts ausgewertet; die Werte von mehrfach vorkommenden Schlüsseln werden überschrieben:

```python
dict1b = {'key1':'value1', 'key2':'value2', 'key3':'value3','key1':'value4'}
dict1b
```

**Möglichkeit 2a:** `dict()` - Liste von Tupels mit Key,Value Paaren

```python
dict2 = dict([('key1','value1'), 
              ('key2','value2'), 
              ('key3','value3')])
dict2
```

**Möglichkeit 2b:** `dict()` - Tupel von Tupels mit Key,Value Paaren

```python
dict2 = dict((('key1','value1'), 
              ('key2','value2'), 
              ('key3','value3')))
dict2
```

**Möglichkeit 2c:** `dict()` - Liste von Listen mit Key,Value Paaren

```python
dict2 = dict([['key1','value1'], 
              ['key2','value2'], 
              ['key3','value3']])
dict2
```

**Spezialfall:** Definition eines Dictionaries, wo der Schlüssel `'key1'` in der Definition mehrfach vorkommt.
<br>
Die Liste oder das Tupel wird von Anfang bis Ende ausgewertet, mehrfach vorkommende Schlüssel werden überschrieben.
<br>
Im Fall unten wird der erste Wert 'value1' des Schlüssels `'key1'` überschrieben.

```python
dict2b = dict([('key1','value1'), 
              ('key2','value2'), 
              ('key3','value3'),
              ('key1','value4')])
dict2b
```

### Löschen eines spezifischen Schlüssel mit zugehörigem Wert 

**Möglichkeit 1:** `del` Befehl

Referenz zu [Python's del statement](https://realpython.com/python-del-statement/)

Neben Funktionen und Methoden gibt es in Python wenige Befehle, die beim Aufruf keine Klammern brauchen,
aber Klammern haben können.
<br>
Referenz: [Why does del (x) with parentheses around the variable name work?](https://stackoverflow.com/questions/39028249/why-does-del-x-with-parentheses-around-the-variable-name-work)

```python
dict1 = {'key1':'value1', 'key2':'value2', 'key3':'value3'}
del dict1['key3']
dict1
```

```python
dict1 = {'key1':'value1', 'key2':'value2', 'key3':'value3'}
del (dict1['key3'])
dict1
```

**Möglichkeit 2** `pop(key)`-Methode - spezifischen Schlüssel aus Dictionary entfernen und Wert zurückgeben

```python
dict1 = {'key1':'value1', 'key2':'value2', 'key3':'value3'}
wert = dict1.pop('key3')
print(wert)
dict1
```

### Dictionary-Methoden 


Methoden sind Funktionen, die sich auf ein Objekt beziehen.
<br>
Schreibweise: **_Objekt.Methode()_**  - Punkt-Notation

Wir definieren uns ein frisches Dictionary:


```python
dict1 = {'key1':'value1', 'key2':'value2', 'key3':'value3'}
dict1
```

#### Rückgabe aller Schlüssel als "Liste" -> `.keys()`-Methode

```python
dict1.keys()
```

#### Rückgabe aller Werte als "Liste" -> `.values()`-Methode

```python
dict1.values()
```

```python
dict1.items()
```

Iterieren über die Schlüssel eines Dictionarys mit der `for`-Schleife

Die `for`-Schleife werden wir noch behandeln.

```python
for key in dict1.keys():
    print(key)
```

```python
for key in dict1.keys():
    print(key,dict1[key])
```

```python
for item in dict1.items():
    print(item, item[0], item[1])
```

### tuple unpacking
Zuweisen der Elemente eines Tupels auf einzelne Variablen: 

```python
(key, value)  = ('key1', 'value1')
print(key)
print(value)
```

Es geht auch ohne Klammern beim Zuweisen:

```python
key, value  = 'key1', 'value1'
print(key)
print(value)
```

Jetzt Einsatz mit der for-Schleife

```python
for key, value in dict1.items():
    print(key,' -> ',value)
```

tuple unpacking muss passen:

- links: zwei Variablen
- rechts: ein Liste mit zwei Elementen

-> je ein Element der Liste wird einer Variable zugeordnet:

```python
key, value  = [('key1', 'value1'),('key2', 'value2')]
print(key)
print(value)
```

```python
for key, value  in [('key1', 'value1'),('key2', 'value2'),('key3', 'value3')]:
    print(key, value)
```

```python
a,b = ['1','2']
print(a)
print(b)
```

hier passt es nicht und es gibt einen `Value`-Fehler:

```python
a,b = ['1','2','3']
```

```python
a,*b = ['1','2','3']
print(a)
print(b)
```
