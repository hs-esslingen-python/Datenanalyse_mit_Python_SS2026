## list comprehension

**Vorteil:** passt in eine Zeile und kann direkt zugewiesen werden 

```
    [ <tue was mit dem Element> for Element in Liste if <bedingung mit dem Element> ]
```


````python
x = [1,2,3,4,5,6,8,9]
````

wir wollen eine Liste mit den Quadraten der Zahlen

### klassische Lösung mit for Schleife

a) Verwendung von `.append()` zum Aufbau der Liste

````python
out = []
for Element in x:
    out.append(Element**2)  # Ein Element wird an die Liste angehängt
    print(f"in der Schleife: {out}")
print(f"am Ende der Schleife: {out}")
````

b) Verwenden von `.extend()` zum Aufbau der Liste:

````python
out = []
for Element in x:
    out.extend([Element**2])  # eine Liste mit einem Element wird angefügt
    print(f"in der Schleife: {out}")
print(out)
````

Um jetzt mit _list comprehension_ in einer Zeile:

````python
[Element**2 for Element in x]
````

````python
out = [Element**2 for Element in x]
out
````

jetzt mit range

````python
out = [Element**2 for Element in range(1,10)]
out
````

Verwendung in einer for-Schleife

````python
for item in [Element**2 for Element in x]:
    print(item)
````

### List Comprehensions mit einer Bedingung

wir wollen eine Liste mit den Quadraten nur der **geraden** Zahlen

````python
out = []
for Element in x:
    if Element%2==0:
        out.append(Element**2)  # Ein Element wird an die Liste angehängt
    print(f"in der Schleife: {out}")
print(f"am Ende der Schleife: {out}")
````

````python
[Element**2 for Element in x if Element%2==0]
````

#### List Comprehensions funktioniert auch mit Dictionaries

```
{<key>:<value tue was mit Element> for Element in Liste if <bedingung mit Element> }
```


````python
x = [1,2,3,4,5,6,8,9]
````

````python
{Element:Element**2 for Element in x }
````

````python
new_dict = {Element:Element**2 for Element in x }
new_dict
````

Verwendung mit einer for-schleife

````python
for k,v in new_dict.items():
    print(f"key:{k} -> value:{v}")
````

````python
for k,v in {Element:Element**2 for Element in x }.items():
    print(f"key:{k} -> value:{v}")
````
