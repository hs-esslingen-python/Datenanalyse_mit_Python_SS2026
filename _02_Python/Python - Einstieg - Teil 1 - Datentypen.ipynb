Python unterscheidet zwischen folgenden Zahlentypen:
   - Ganzzahlen (Integer) -> Zahlen ohne Dezimaltrenner `.` z.B.: 5
   - Fließkommazahlen (Floatingpointzahlen):
       - Zahlen mit Dezimaltrenner `.`  z.B.: 3.1415 oder  3.0
       - Zahlen mit Exponentschreibweise z.B.: 1e10
   
Die Addition (`+`), die Subtraktion (`-`) und die Multiplikation (`*`) von zwei Integerzahlen, liefert wieder eine Integerzahl.

Die Division (`/`) von zwei Integern liefert immer eine Fließkomma-Zahl.

Mit der [type()](https://www.geeksforgeeks.org/python-type-function/)-Funktion wird der "Typ" eines Python Objekts also auch der einer Zahl ([Literal](https://de.wikipedia.org/wiki/Literal)) bestimmt.

- `int`: Integer - Ganzzahl
- `float`: Fließkommanzahl

Ausblick: Die `type`-Funktion wird bei _Mutable und Immutable Objects_ wieder aufgegriffen.

## ganzzahlige Division und Division mit Rest (modulo)

- [floor division](https://python-reference.readthedocs.io/en/latest/docs/operators/floor_division.html) `//` ganzzahlige Division mit "Abrundungsfunktion"
- [modulus](https://python-reference.readthedocs.io/en/latest/docs/operators/modulus.html) `%` "Rest"

Es gilt folgender Zusammenhang zwischen Quotient, Dividend, Divisior, Zähler und Nenner; vergleiche [Division - Wikipedia](https://de.wikipedia.org/wiki/Division_(Mathematik)):
$$
    Quotient = \frac{\text{Dividend}}{\text{Divisor}} = \frac{\text{Zähler}}{\text{Nenner}}
$$

Die "_Floor division_" `//` (ganzzahlige Division mit "Abrundungsfunktion") ist definiert als:
$$
    \text{Zähler} ~ \text{//} ~ \text{Nenner} = \text{floor}(\text{Zähler} ~ \text{/} ~ \text{Nenner})
$$
mit $\text{floor}()$  als der Abrundungsfunktion.

Die Abrundungsfunktion  $\text{floor}(x)$ liefert die größte Integer, die kleiner oder gleich der Zahl x im Argument ist:
- $\text{floor}()$ für positive Argumente: $\text{floor}(3.2)$ = 3 (dies entspricht dem Abschneiden der Nachkommanstellen)
- $\text{floor}()$ für negative Argumente: $\text{floor}(-3.2)$ = -4 (abrunden heißt in Richtung minus unendlich gehen)

Die Division mit Rest, vergleiche [Division mit Rest - Wikipedia](https://de.wikipedia.org/wiki/Division_mit_Rest) oder
[Python Floor Division - pythontutorial.net](https://www.pythontutorial.net/advanced-python/python-floor-division) wird durchfolgende Gleichung beschrieben:

$$
\begin{align}
    \text{Zähler} &= \text{Nenner} ~ * ~ \underbrace{(\text{Zähler} ~ \text{//} ~ \text{Nenner})}_{\text{floor division}} ~+~ \underbrace{(\text{Zähler}~\text{\%}~\text{Nenner})}_{\text{modolo}} \\
    \text{Z} &= \text{N} ~ * ~\underbrace{(\text{Z} ~ \text{//} ~ \text{N})}_{\text{floor division}}~+~ \underbrace{(\text{Z}~\text{\%}~\text{N})}_{\text{modolo}}   
\end{align}
$$
mit $\text{Z}$ für Zähler und $\text{N}$ für Nenner.

Das Ergebnis der Division ist der ganzzahlige Anteil der Division. Falls beide Operanden Integerwerte sind, stellt auch das Ergebnis einen Integerwert dar. Falls jedoch mindestens einer der Operanden ein Floatwert ist, so wird der ganzzahlige Anteil der Division als Float ausgegeben.

**Beispiel - nur positive Zahlen**: 5 durch 2 ist 2 Rest 1

```python
5//2
```

```python
type(5//2)
```

```python
5%2 
```

```python
type(5%2)
```

Was passiert bei negativen Zahlen?

**Beispiel - mit einer negativen Zahl**: -5 durch 2 ist -3 Rest 1

```python
-5//2
```

```python
-5%2
```

Nebenrechnung zu "Beispiel - mit einer negativen Zahl":

Es gilt die Formel:
$$
    \text{Z} = \text{N} ~ * ~ (\text{Z} ~ \text{//} ~ \text{N}) ~+~ (\text{Z}~\text{\%}~\text{N})   
$$
mit $\text{Z}=-5$ und $\text{N}=2$.

Die float Division ist:
$$
    \text{Z} ~ \text{/} ~ \text{N} = -5  ~ \text{/} ~ 2= -2.5 
$$
Abgerunden ergibt dies -3; dies ist die floor division:
$$
    \text{Z} ~ \text{//} ~ \text{N} = \text{-5} ~ \text{//} ~ \text{2} = -3
$$

Dann ergibt sich durch Umstellen der oben genannten Formel für die Modulo Operation $(\text{Z}~\text{\%}~\text{N})$:
$$
\begin{align}
    \text{Z}~\text{\%}~\text{N}  &= \text{Z}- \text{N} ~ * ~ (\text{Z} ~ \text{//} ~ \text{N}) \\
    \text{Z}~\text{\%}~\text{N}  &= (-5) - (2) ~ * ~ (-3) \\
    \text{Z}~\text{\%}~\text{N}  &= (-5) - (-6) \\
    \text{Z}~\text{\%}~\text{N}  &= 1   
\end{align}
$$

#### Was liefert die floor division bei realen Zahlen (float)?

Das Ergebnis ist eine Float-Zahl, sobald einer der Operanden eine Float-Zahl ist.

```python
5//2.0
```

```python
type(5//2.0)
```

```python
5.0//2
```

```python
type(5.0//2)
```

Was passiert wenn wir keine ganzen Zahlen verwenden?

```python
5//2.1
```

```python
5%2.1
```

```python
print(">0123456789<")
print(f">{5%2.1:10.2f}<")
```

```python
5/2.1
```

Es gilt die Formel:
$$
    \text{Z} = \text{N} ~ * ~ (\text{Z} ~ \text{//} ~ \text{N}) ~+~ (\text{Z}~\text{\%}~\text{N})   
$$
mit $\text{Z}=5$ und $\text{N}=2.1$.

Die float Division ist:
$$
    \text{Z} ~ \text{/} ~ \text{N} = 5  ~ \text{/} ~ 2.1= 2.380952380952381
$$
Abgerunden ergibt dies -3; dies ist die floor division:
$$
    \text{Z} ~ \text{//} ~ \text{N} = \text{5} ~ \text{//} ~ \text{2.1} = 2
$$

Dann ergibt sich durch Umstellen der oben genannten Formel für die Modulo Operation $(\text{Z}~\text{\%}~\text{N})$:
$$
\begin{align}
    \text{Z}~\text{\%}~\text{N}  &= \text{Z}- \text{N} ~ * ~ (\text{Z} ~ \text{//} ~ \text{N}) \\
    \text{Z}~\text{\%}~\text{N}  &= (5) - (2.1) ~ * ~ (2) \\
    \text{Z}~\text{\%}~\text{N}  &= (5) - (4.2) \\
    \text{Z}~\text{\%}~\text{N}  &= 0.8   
\end{align}
$$

## Potenzieren

Potenzieren: Basis `**` Exponent

```python
2**8
```

```python
type(2**8)
```

**Anmerkung:**

Das Potenzieren (`**`) von zwei Integerzahlen, liefert wieder eine Integerzahl.


Das Ergebnis ist eine Float-Zahl, sobald einer der Operanden eine Float-Zahl ist.

```python
type(2.0**8)
```

```python
type(2**8.0)
```

## "Punkt vor Strichrechnung"

Bei der Bearbeitung eines Ausdruck gilt "Punkt vor Strickrechnung".
Multiplikation und Division wird vor Addition und Subtraktion ausgewertet.
Durch runde Klammern `()` kann die Reihenfolge aber eindeutig beschrieben werden.

```python
2+3*4+5
```

```python
(2+3)*(4+5)
```
