## ganzzahlige Division und Division mit Rest (modulo)

- [floor division](https://python-reference.readthedocs.io/en/latest/docs/operators/floor_division.html) `//` ganzzahlige Division mit "Abrundungsfunktion"
- [modulus](https://python-reference.readthedocs.io/en/latest/docs/operators/modulus.html) `%` "Rest"

Es gilt folgender Zusammenhang zwischen Quotient, Divident, Divisior, Zähler und Nenner; vergleiche [Division - Wikipedia](https://de.wikipedia.org/wiki/Division_(Mathematik)):
$$
    Quotient = \frac{\text{Dividend}}{\text{Divisor}} = \frac{\text{Zähler}}{\text{Nenner}}
$$

Die "**Floor division**" `//` (ganzzahlige Division mit "Abrundungsfunktion") ist definiert als:
$$
    \text{Zähler} ~ \text{//} ~ \text{Nenner} = \text{floor}(\text{Zähler} ~ \text{/} ~ \text{Nenner})
$$
mit $\text{floor}()$  als der Abrundungsfunktion.

Die Abrundungsfunktion  $\text{floor}()$ liefert die größte Integer, die kleiner oder gleich der Zahl im Argument ist:
- $\text{floor}()$ für positive Argumente: $\text{floor}(3.2)$ = 3 (dies entspricht dem Abschneiden der Nachkommanstellen)
- $\text{floor}()$ für negative Argumente: $\text{floor}(-3.2)$ = -4 

Die **Division mit Rest** (**modulo**) wird durch folgende Gleichung beschrieben:

$$
\begin{align}
    \text{Zähler} &= \text{Nenner} ~ * ~ \underbrace{(\text{Zähler} ~ \text{//} ~ \text{Nenner})}_{\text{floor division}} ~+~ \underbrace{(\text{Zähler}~\text{\%}~\text{Nenner})}_{\text{modulo}} \\
    \text{Z} &= \text{N} ~ * ~\underbrace{(\text{Z} ~ \text{//} ~ \text{N})}_{\text{floor division}}~+~ \underbrace{(\text{Z}~\text{\%}~\text{N})}_{\text{modulo}}   
\end{align}
$$
mit $\text{Z}$ für Zähler und $\text{N}$ für Nenner; vergleiche [Division mit Rest - Wikipedia](https://de.wikipedia.org/wiki/Division_mit_Rest) oder
[Python Floor Division - pythontutorial.net](https://www.pythontutorial.net/advanced-python/python-floor-division).

Das Ergebnis der Division ist der ganzzahlige Anteil der Division. Falls beide Operanden Integerwerte sind, stellt auch das Ergebnis einen Integerwert dar. Falls jedoch mindestens einer der Operanden ein Floatwert ist, so wird der ganzzahlige Anteil der Division als Float ausgegeben.

## Beispiele:

| Z | N | Z//N | Z%N |
|--:|--:|-----:|----:|
| 10|  3|    3 |    1|
| 10| -3|   -4 |   -2|
|-10|  3|   -4 |    2|
|-10| -3|    3 |   -1|



----
**Beispiel 1**: 10 durch 3 

Ergebnis: 
   - floor division 10//3 = 3 und 
   - modulo 10%3 = 1

```python
10//3
```

```python
10%3
```

Es gilt die Formel:
$$
    \text{Z} = \text{N} ~ * ~ (\text{Z} ~ \text{//} ~ \text{N}) ~+~ (\text{Z}~\text{\%}~\text{N})   
$$
mit $\text{Z}=10$ und $\text{N}=3$.

Die float Division ist:
$$
    \text{Z} ~ \text{/} ~ \text{N} = 10  ~ \text{/} ~ 3= 3.333 
$$
Abgerundet ergibt dies 3; also ist die floor division:
$$
    \text{Z} ~ \text{//} ~ \text{N} = 10 ~ \text{//} ~ 3 = 3
$$

Dann ergibt sich durch Umstellen der Formel für die Modulo Operation $(\text{Z}~\text{\%}~\text{N})$:
$$
\begin{align}
    \text{Z}~\text{\%}~\text{N}  &= \text{Z}- \text{N} ~ * ~ (\text{Z} ~ \text{//} ~ \text{N}) \\
    \text{Z}~\text{\%}~\text{N}  &= 10 - 3 ~ * ~ 3 \\
    \text{Z}~\text{\%}~\text{N}  &= 10 - 9 \\
    \text{Z}~\text{\%}~\text{N}  &= 1   
\end{align}
$$

**Beispiel 2**: 10 durch -3  

Ergebnis:
- floor division 10//-3 = -4 und 
- modulo  10%-3 = -2

```python
10%-3
```

Es gilt die Formel:
$$
    \text{Z} = \text{N} ~ * ~ (\text{Z} ~ \text{//} ~ \text{N}) ~+~ (\text{Z}~\text{\%}~\text{N})   
$$
mit $\text{Z}=10$ und $\text{N}=-3$.

Die float Division ist:
$$
    \text{Z} ~ \text{/} ~ \text{N} = 10  ~ \text{/} ~ (-3)= -3.333 
$$
Abgerundet ergibt dies -4; also ist die floor division:
$$
    \text{Z} ~ \text{//} ~ \text{N} = 10 ~ \text{//} ~ (-3) = -4
$$

Dann ergibt sich durch Umstellen der Formel für die Modulo Operation $(\text{Z}~\text{\%}~\text{N})$:
$$
\begin{align}
    \text{Z}~\text{\%}~\text{N}  &= \text{Z}- \text{N} ~ * ~ (\text{Z} ~ \text{//} ~ \text{N}) \\
    \text{Z}~\text{\%}~\text{N}  &= 10 - (-3) ~ * ~ (-4) \\
    \text{Z}~\text{\%}~\text{N}  &= 10 - 12 \\
    \text{Z}~\text{\%}~\text{N}  &= -2   
\end{align}
$$
