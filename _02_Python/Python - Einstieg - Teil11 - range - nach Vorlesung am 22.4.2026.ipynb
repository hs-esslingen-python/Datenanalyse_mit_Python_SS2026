## range()

erzeugt ein Objekt, welches eine Zahlenfolge erzeugt

- [`range(stop)`](https://docs.python.org/3/library/functions.html#func-range) ->  Zahlenfolge von 0 bis stop-1  (stop ist nicht enthalten - wie beim Slicing)
- [`range(start, stop)`](https://docs.python.org/3/library/functions.html#func-range)  -> Zahlenfolge startet bei start
- [`range(start, stop, step)`](https://docs.python.org/3/library/functions.html#func-range) -> zusätzlich Angabe einer Schrittweite



```python
range(5)
```

```python
type(range(5))
```

[`range()`](https://docs.python.org/3/library/functions.html#func-range) ist lazy.
<br>
Zahlenfolge Objekt in eine Liste wandeln

```python
list(range(5))
```

Schleifen zum Zählen von 0 bis 4

```python
for i in range(5):
    print(i)
```

Wenn die Iterationvariable nicht genutzt wird kann sie durch ein Unterscore `_` ersetzt werden 

```python
for _ in range(5):
    print("Action")
```

Schleife über die Zahlen 10 bis 50 in 10er Schritten

```python
for i in range(10,51,10):
    print(i)
```

Schleife über die Zahlen 50 bis 10 ins 10er Schritten

```python
for i in range(50,9,-10):
    print(i)
```

Schleife über die Zahlen -40 bis +40 ins 10er Schritten

```python
for i in range(-40,41,10):
    print(i)
```

Schleife über die Zahlen 40 bis -40 ins 10er Schritten

```python
for i in range(40,-41,-10):
    print(i)
```
