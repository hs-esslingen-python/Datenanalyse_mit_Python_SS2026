### Lösungs-Idee:

In einer Schleife gehen wir die Zeichenkette Zeichen für Zeichen durch.

Die Variable `Gedicht` ist eine Zeichkette, ein _Iterierbarer Typ_, dessen einzelne Zeichen nacheinander mit einer `for`-Schleife geholt werden können:

````python
for Zeichen in Gedicht:
    print(Zeichen, end='')
````

Der Text enthält große und kleine Buchstaben. Wir wollen bei der Statistik hier aber nicht unterscheiden.
Mit der Funktion [`.lower()`](https://docs.python.org/3/library/stdtypes.html#str.lower) läßt sich eine Zeichenkette oder ein einzeln Zeichen in ein kleine Zeichen umwandeln:

````python
for Zeichen in Gedicht:
    Zeichen = Zeichen.lower()
    print(Zeichen, end='')
````

Der Text enthält auch noch **Interpunktionszeichen**: Punkt, Komma; Semikolon und Ausrufezeichen.
Wir wollen nur die Buchstaben von 'a' bis 'z' in der Statistik haben.
Dazu schreiben wir eine Funktion, der wir ein Zeichen übergeben und die ``True` zurückgibt, wenn das Zeichen ein Buchstabe von 'a' bis 'z' ist.
<br>
Jetzt gibt es auch noch Sonderzeichen 'ß'.

````python
def Zeichen_ist_Buchstabe(Zeichen):
    if Zeichen >= 'a' and Zeichen <= 'z' or Zeichen in ['ß', 'ä', 'ö','ü']:
        return True
    return False
````

````python
for Zeichen in Gedicht.lower():
    if Zeichen_ist_Buchstabe(Zeichen):
        print(Zeichen, end='')
````

### Alternative Ansatz:

Mit der Methode [`.lower()`](https://docs.python.org/3/library/stdtypes.html#str.lower) wandeln wir das jeweilige Zeichen in ein klein geschriebenes Zeichen um.

Mit der Methode Methode [`.isalpha()`](https://docs.python.org/3/library/stdtypes.html#str.isalpha) stellen wir sicher, dass es ein Buchstabe oder eine Zahl aber kein Leerzeichen ist.

Zum Zählen der Häufigkeit der Zeichen legen wir uns ein Python Dictionary an, wo die Schlüssel 'keys' die Buchstaben und die zugehörige Werte die Zähler darstellen.

Das Python Dictionary wird vor der Schleife als leeres Dictionary initialisiert und dann dynamische in der Schleife um die auftretenden Zeichen erweitert und auf die aktuellen Zählerstände gebracht.

Während des Schleifendurchlauf der Zeichenkette, wir zunächst geprüft, ob der treffende Buchstabe schon als Schlüssel im Dictionary angelegt wurde.
Wenn der Schlüssel nicht vorhandnen, wird der Buchstabe als Schlüssel angelegt und mit dem Wert Null initialsiert.

Dann wird der Wert des Schlüssels, um eins nach oben gezählt.


````python
Statistik = {}                                 # leeres Dictionary anlegen
for b in Gedicht:                              # Schleife für die einzelnen Zeichen der Zeichenkette
    b = b.lower()                              # Großbuchstaben in Kleinbuchstaben wandeln
    if b.isalpha():                            # nur Zählen bei Buchstaben
        if not b in Statistik:                 # Prüfen, ob Buchstabe als Schlüssel in Dictionary vorhanden
            Statistik[b] = 0                   # Wenn nicht dann anlegen
        Statistik[b] += 1                      # Zahlen für den Buchstaben hochzählen
        
        # Alternative Implementierung
        #if b in Statistik:
        #    Statistik[b] +=1
        #else:
        #    Statistik[b] = 1
        #print(b)
Statistik                                      # Ausgabe des Dictionaries
````

Aus dem Dictionary eine Liste mit Tupeln anlegen, wo der ersten Eintrag der Buchstabe und der zweite Eintrag der Zählwert ist.

````python
Liste = []                                       # leere Liste anlegen
for key in Statistik.keys():                     # Schleife für alle Schlüssel des Dictionaries
    Liste.append( (key,Statistik[key]) )         # Liste um ein Tupel erweitern
Liste
````

### Sortieren
Jetzt wollen wir die Liste nach fallenden Häufigkeit sortieren

````python
sorted(Liste)
````

Die Liste wurde nach den Buchstaben sortiert. Dies ist nicht was wir wollen.

Jedoch bietet die [`sorted()`](https://docs.python.org/3/library/functions.html#sorted)-Funktion mit dem Parameter `key` die Möglichkeit eine Funktion anzugeben, mit der der Vergleich bei Sortierung bestimmt wird.

Die Liste, die wir Sortieren wollen, besteht aus Tupeln, wobei der zweite Eintrag, die Anzahl, für das Sortieren entscheidend ist.
Wir schreiben uns eine Funktion, die den zweiten Eintrag aus dem Tupel zurückgibt:

````python
def get_counter(item):
    return item[1]
````

````python
get_counter(('b', 3))
````

Verwenden der Python Funktion `get_counter()` beim Sortieren.
Bitte beachten: Die Python Funktion wird ohne runde Klammern übergeben, weil wir die Funktion als Objekt übergeben wollen. Die runden Klammern würden ein Ausführen der Funktion bewirken.

````python
sorted(Liste, key=get_counter)
````

Das sieht schon besser aus.
Jedoch stimmt noch nicht die Richtung der Suche.
Wir haben die Werte in aufsteigender Reihenfolge, wollen sie aber in absteigender Reihenfolge haben.
Dazu nutzen wird den Parameter `reverse=True` der `sorted()`-Funktion.

````python
sorted(Liste, key=get_counter, reverse=True)
````

Das ist das was wir haben wollen

Jetzt noch das gleich jedoch mit der integrierten lambda Funktion:

````python
sorted(Liste, key=lambda item: item[1], reverse=True)
````

Statt `Liste` können wir auch `Statistik.items()` verwenden:

````python
sorted(Statistik.items(), key=lambda item: item[1], reverse=True)
````

## Integration in eine Funktion
Jetzt das ganze als eine Python Funktion:

````python
def Buchstaben_Analyse(Text):
    """ Funktion bestimmt die Anzahl der Buchstaben in Text
        und gibt eine Liste mit Tupel (Buchstabe, Anzahl) in absteigender
        Reihenfolge der Häufigkeit aus.
    """
    Text = Text.lower()                            # Großbuchstaben in Kleinbuchstaben wandeln
    Statistik = {}                                 # leeres Dictionary anlegen
    for b in Text:                                 # Schleife für die einzelnen Zeichen der Zeichenkette
        if b.isalpha():                            # nur Zählen bei Buchstaben
            if not b in Statistik:                 # Prüfen, ob Buchstabe als Schlüssel in Dictionary vorhanden
                Statistik[b] = 0                   # Wenn nicht dann anlegen
            Statistik[b] += 1                      # Zahlen für den Buchstaben hochzählen
    return sorted(Statistik.items(), key=lambda item: item[1], reverse=True)
````

````python
Buchstaben_Analyse(Gedicht)
````

## Zusatzaufgaben: Umwandlung des 'ß' in 'ss'

Dazu können wir die Methode [`.replace()`](https://docs.python.org/3/library/stdtypes.html#str.replace) nutzen.
Die  Methode [`.replace()`](https://docs.python.org/3/library/stdtypes.html#str.replace) wird auf eine Zeichenkette angewendet, da erste Argument ist die Teilzeichenkette, die in die Zeichenkette im zweiten Argument übergeben wird.

````python
"Spaß".replace("ß","ss")
````

````python
def Buchstaben_Analyse(Text):
    """ Funktion bestimmt die Anzahl der Buchstaben in Text
        und gibt eine Liste mit Tupel (Buchstabe, Anzahl) in absteigender
        Reihenfolge der Häufigkeit aus.
    """
    Text = Text.replace("ß","ss")                  # Umwandlung des 'ß' in 'ss'
    Text = Text.lower()                            # Großbuchstaben in Kleinbuchstaben wandeln
    
    Statistik = {}                                 # leeres Dictionary anlegen
    for b in Text:                                 # Schleife für die einzelnen Zeichen der Zeichenkette
        
        if b.isalpha():                            # nur Zählen bei Buchstaben
            if not b in Statistik:                 # Prüfen, ob Buchstabe als Schlüssel in Dictionary vorhanden
                Statistik[b] = 0                   # Wenn nicht dann anlegen
            Statistik[b] += 1                      # Zahlen für den Buchstaben hochzählen
    return sorted(Statistik.items(), key=lambda item: item[1], reverse=True)
````

````python
Buchstaben_Analyse(Gedicht)
````
