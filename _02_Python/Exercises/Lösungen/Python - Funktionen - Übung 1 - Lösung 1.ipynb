## Buchstaben Statistik

````python
Gedicht = "Fest gemauert in der Erden"\
          "Steht die Form aus Lehm gebrannt."\
          "Heute muß die Glocke werden! "\
          "Frisch, Gesellen, seid zur Hand! "\
          "Von der Stirne heiß "\
          "Rinnen muß der Schweiß, "\
          "Soll das Werk den Meister loben; "\
          "Doch der Segen kommt von oben."
````

````python
Gedicht
````

-----
**Aufgabe**: Erstelle eine Statistik über die Buchstaben im Gedicht. 
Gebe das Ergebnis als Liste als Paare ('Buchstabe', Anzahl) geordnet mit abfallender Häufigkeit.

Hinweis: Verwende die Funktion `sorted()` zum Sortieren.

Ergebnis:
```
[('e', 35),
 ('n', 17),
 ('r', 15),
 ('d', 13),
 ('s', 12),
 ('i', 10),
 ('t', 9),
 ('o', 9),
 ('m', 8),
 ('h', 8),
 ('l', 7),
 ('u', 6),
 ('g', 5),
 ('a', 5),
 ('ß', 4),
 ('c', 4),
 ('f', 3),
 ('b', 3),
 ('k', 3),
 ('w', 3),
 ('v', 2),
 ('z', 1)]
 ```


````python
#solution
Statistik = {}  # dictionry
for c in Gedicht:
    #print(c)
    if c.isalpha():
        c=c.lower()
        if not c in Statistik:
            Statistik[c] = 0
        Statistik[c] +=1
sorted(Statistik.items(), key=lambda item: item[1], reverse=True)
````

# Lösung mit Pandas

````python
import pandas as pd
````

````python
liste_der_buchstaben = [*Gedicht]
liste_der_buchstaben
````
