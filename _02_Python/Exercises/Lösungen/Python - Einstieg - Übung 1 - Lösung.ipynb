## Ganzzahldisvision und Modulo
Schreibe eine Funktion die eine Divsion mit Rest berechnet und auf dem Bildschirm ausgibt

Berechne dann das Beispiel 30 durch 7 teilt.

Verwende zur Ausgabe einen eine formatierte Ausgabe.

Ergebnis: `30 geteilt durch 7 ist 4 mit Rest 2`

Wikipedia: 
- [Division](https://de.wikipedia.org/wiki/Division_(Mathematik))
- [Division_mit_Rest](https://de.wikipedia.org/wiki/Division_mit_Rest)

Python
- [modulus](https://python-reference.readthedocs.io/en/latest/docs/operators/modulus.html)
- [floor division](https://python-reference.readthedocs.io/en/latest/docs/operators/floor_division.html)

````python
#solution
# format
def Division_mit_Rest(Dividend,Divisor):
    Ganzzahlquotient=Dividend//Divisor
    Rest=Dividend%Divisor
    print("{} geteilt durch {} ist {} mit Rest {}".format(Dividend,Divisor,Ganzzahlquotient,Rest))
````

````python
Division_mit_Rest(30,7)
````

````python
#solution
# f-string
def Division_mit_Rest2(Dividend,Divisor):
    print(f"{Dividend} geteilt durch {Divisor} ist {Dividend//Divisor} mit Rest {Dividend%Divisor}")
````

````python
Division_mit_Rest2(30,7)
````

### "Runden": Fließkommazahl in Integerzahl wandeln

Berechne für folgende Zahlen: [-3.2, -2.7, 1.7, 3.2] folgende Werte:
* int(x)   - Abschneiden der Nachkommastellen
* round(x) - zur näherliegende Integer runden
* round(x,2) - auf zwei Stellen genau runden
* math.floor(x) - nehme die kleinere Integerzahl 
* math.ceil(x)  - nehme die größere Integerzahl
* math.trunc(x) - Nachkommastelle abtrennen

### Quadratwurzel
Schreibe eine Funktion `Betrag_Vektor(x,y,z)` die den Betrag eines 3D-Vektors berechnet und berechne den Betrag des Vektors (x,y,z) =(1,2,3)

````python
#solution
import math
def Betrag_Vektor(x,y,z):
    return math.sqrt(x**2+y**2+z**2)
````

````python
Betrag_Vektor(1,2,3)
````

Erweitere die Funktion `Betrag_Vektor(x,y,z)`, so dass ein Ausruf für 2D Vektoren möglich wird: `Betrag_Vektor(x,y)` und teste den Aufruf `Betrag_Vektor(3,4)`

Hinweis: default arguments

````python
#solution
def Betrag_Vektor(x,y,z=0):
    return math.sqrt(x**2+y**2+z**2)
````

````python
Betrag_Vektor(3,4)
````

Erweitere die Funktion um den Betrag von Vektoren beliebiger Dimension berechnen zu können

Hinweis: positional arguments

````python
#solution
def Betrag_Vektor(*args):
    Summe_der_Quadrate = 0
    for arg in args:
        Summe_der_Quadrate += arg**2
    
    return math.sqrt(Summe_der_Quadrate)
````

````python
Betrag_Vektor(1,2,3)
````

````python
Betrag_Vektor(3,4)
````

````python
Betrag_Vektor(1,2,3,4,5,6)
````

Erweitere auch um keyword Argumente

````python
#solution
def Betrag_Vektor(*args,**kwargs):
    Summe_der_Quadrate = 0
    for arg in args:
        Summe_der_Quadrate += arg**2
    for kwarg in kwargs.values():
        Summe_der_Quadrate += kwarg**2
    
    return math.sqrt(Summe_der_Quadrate)
    
````

````python
Betrag_Vektor(x=1,y=2,z=3)
````

Was macht die folgende Funktion?

Hinweis: Nutze die Zeichenkette "Wenn Fliegen hinter Fliegen fliegen, fliegen Fliegen nach"

````python
def Unbekannt(a, b):
    k = 0
    for c in b.lower().replace(",", "").split():
        if c == a:
            k +=1
    return k 
````

````python
Unbekannt('fliegen','Wenn Fliegen hinter Fliegen fliegen, fliegen Fliegen nach')
````

````python
#solution
'Wenn Fliegen hinter Fliegen fliegen, fliegen Fliegen nach'.lower().replace(",", "").split()
````
