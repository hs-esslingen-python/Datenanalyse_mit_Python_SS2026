## Übung zu Listen

Aufgabe1: Erstellen Sie eine Liste _Farben1_, die die Elemente 'rot', 'grün' und 'blau' enthält.
Nutzen Sie eckige Klammer `[]`

Aufgabe2: Erzeugen Sie eine zweite Liste _Farben2_ mit den Elemente 'gelb', 'cyan', 'magenta' indem Sie zuerst ein Tupel erstellen und dann mit der `list()`-Funktion in eine Liste wandeln.

Aufgabe3: Fassen Sie die beiden Listen zusammen in einer Liste _Farben3_ zusammen. 

Aufgabe4: Erstellen Sie ein Referenz der Liste und nennen Sie diese _Farben1_Referenz_

Aufgabe5: Erstellen Sie eine Kopie der Liste _Farben1_ und nennen Sie sie _Farben1_Kopie_

Aufgabe6: Fügen Sie die Elemente der Liste _Farben2_ in die Liste _Farben1_ hinein.

Aufgabe7: Geben Sie Liste _Farben1_Referenz_ und Liste _Farben1_Kopie_ aus

Aufgabe8: Hängen Sie das Element 'violett' an das Ende der Liste _Farben3_.

Aufgabe9: Wie viele Element enthält die Liste _Farben3_?


Machen Sie Beispiele zu den folgenden Methoden 
+ `.insert()`- Methode:  einfügen an beliebiger Position
+ `.remove()`- Methode:  entfernen eines Elements aus der Liste anhand des Elementnamens
+ `.pop()` - Methode:    letzten Elements aus einer Liste entfernen und zurückgeben
+ `.reverse()` - Methode: Reihenfolge der Liste umkehren
+ `.sort()` - Methode:   Sortieren der Liste 
+ `.index()` - Method:   Index zu einem gegeben Element suchen
