### Datentyp Sets - Mengen 

_Sets_ sind ungeordnete Ansammlung von Werte ohne Doppelungen.
<br>
_Sets_ können nur immutable Objekte enthalten.

[Mengen in Python](https://www.data-science-architect.de/mengen-python/)

In Python werden _Sets_ mit gescheiften Klammer `{}` definiert. 

```python
{1,2,3}
```

Wir definieren ein _Set_ mit vielen doppelten Elementen:

```python
{1,2,3,1,2,1,2,3,1,3,2,2,2,2,1,1,2}
```

Jetzt wollen wir bestimmen welchen Datentyp Python für einen _Set_ zurückgibt.

```python
menge = {1,2,3,1,2,1,2,3,1,3,2,2,2,2,1,1,2}
type(menge)
```

Verwendung der Funktion `set` um z.B. aus einer Liste ein Set zu definieren.

```python
set([1,2,3])
```

Naheliegend, ein _Set_ hat den Datentyp `set`.

Jetzt wollen wir prüfen, ob ein Wert in einem Set enthalten ist.
<br>
Hierzu verwenden wir den `in` Operator. 
<br>
Der folgende Ausdruck <Wert> in <Set> liefert ein Boolean zurück:
<br>
a) `True` wenn der Wert im Set vorhanden ist

```python
1 in menge
```

b) `False` wenn der Wert nicht im Set vorhanden ist.

```python
4 in menge
```

Set mit Elementen unterschiedlicher Datentypen, aber alle Elemente sind **immutable** Objekte.

```python
import math
{2,1,5,0,2,4,3,4,'bc','a',(1,2,3), math.pi,(1,2,3)}
```

### Sets dürfen keine mutable Objekte verwenden.
Wir bekommen eine `TypeError`-Fehlermeldung, wenn wir ein mutable Objekt verwenden.
<br>
Im folgenden Beispiel ist die Liste `[1,2,3]` ein mutable Objekt.

```python
{1,2,3,4,'a','bc',[1,2,3]}
```

### Verknüpfung von Mengen:

Wir definieren zwei Mengen: _Menge1_ und _Menge2_

```python
Menge1 = {1,2,3,4}
Menge2 =     {3,4,5,6}
```

#### Schnittmenge: `.intersection()` - Methode

Die Schnittmenge ist die Menge der Elemente, die in beiden Mengen vorkommen.

```python
Menge1.intersection(Menge2)
```

#### Vereinigungsmenge: `.union()` - Methode

Die Vereingungsmenge ist die Menge aller Elemente der beiden Mengen

```python
Menge1.union(Menge2)
```

#### Differenzmenge: `.difference()` -  Methode

Die Differenzmenge ist die Menge der _Menge1_ ohne die Elemente, die auch in _Menge2_ vorkommen.

```python
Menge1.difference(Menge2)
```

```python
Menge2.difference(Menge1)
```

-----
Sets selber sind **mutable**, d.h. sie behalten ihre Identifikationsnummer auch wenn sie verändert werden

Mit der `add()`-Methode kann einem Set ein Element hinzugefügt werden:

```python
set1 = {1,2,3}
print('Inhalt',set1)
print('id()',id(set1))
```

Element 4 hinzufügen

```python
set1.add(4)
print('Inhalt',set1)
print('id()',id(set1))
```

Element 5 hinzufügen

```python
set1.add(5)
print('Inhalt',set1)
print('id()',id(set1))
```

Die Anzahl der Element in einer Menge kann über die Funktion `len()` bestimmt werden:

```python
len(set1)
```

### Zugriff auf die Element mit eckigen Klammern `[]` wie bei Listen nicht möglich

Bei Menge (Sets) gibt es keine Reihenfolge der Elemente

```python
set1[0]
```
