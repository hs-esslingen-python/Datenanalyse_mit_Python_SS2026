## Mutable und Immutable Objects

In  Python werden sog. Mutable- (veränderliche) von Immutable- (unveränderliche) Objects unterschieden, siehe 
[Mutable und Immutable Objects](https://www.data-science-architect.de/mutable-und-immutable-objects/).

Zu den **Immutable**-Objekten zählen:

    int, float, string, bool, decimal, complex
    tuple
    range
    frozenset
    bytes
    
Variablen, denen ein Immutable-Objekt zugewiesen wurde, können überschrieben werden. Jedoch wird dann das Immutable-Objekt durch ein neues Objekt ersetzt.

Zu den **Mutable**-Objekten zählen:

    list
    dictionary
    set
    User-Definierte Objekte

- `type()` [type()](https://www.geeksforgeeks.org/python-type-function/)-Funktion bestimmt den "Typ" eines Python Objekts.
- `id()` [id()](https://www.programiz.com/python-programming/methods/built-in/id)-Funktion bestimmt die eindeutige "Identikationsnummer" eines Python Objekts

### Datentyp Integer - Ganzzahl 

```python
Ganzzahl = 5
type(Ganzzahl), id(Ganzzahl)
```

```python
Ganzzahl = 6
type(Ganzzahl), id(Ganzzahl)
```

Beim Überschreiben ändert sich die Identifikationsnummer des Objekts `Ganzzahl`.

### Datentyp Float - Fließkommazahl

```python
Fliesskommazahl = 5.5
type(Fliesskommazahl), id(Fliesskommazahl)
```

```python
Fliesskommazahl = 5.6
type(Fliesskommazahl), id(Fliesskommazahl)
```

Beim Überschreiben ändert sich die Identifikationsnummer des Objekts `Fliesskommazahl`.

### Datentyp String - Zeichenketten

```python
String = 'Hallo'
type(String),id(String)
```

```python
String = 'Hallo2'
type(String),id(String)
```

Beim Überschreiben ändert sich die Identifikationsnummer des Objekts `String`.

### Datentyp Boolean  - mit den Werten `True` `False`

```python
boolean = True
type(boolean), id(boolean)
```

```python
boolean = False
type(boolean), id(boolean)
```

```python
boolean2 = True
type(boolean2), id(boolean2)
```

```python
boolean2 = False
type(boolean2), id(boolean2)
```

Anmerkung: Das Objekt `True` und das Objekt `False` gibt es nur einmal. Dies ist anhand der Identifikationsnummer erkennbar.

True and False nach Integer gewandelt:

```python
print(f" True ist nach Integer gewandelt: {int(True)}")
print(f" False ist nach Integer gewandelt: {int(False)}")
```

### Datentyp Liste 

den Datentyp "Liste" werden wir im folgenden noch ausführlich behandeln.
<br>
Eine Liste soll hier aber als Beispiel für ein **Mutable**-Objekt genommen werden. 

Schritt 1: Liste neu anlegen

```python
Liste = [1,2,3]
print(Liste)
type(Liste), id(Liste)
```

Schritt 2: an das Ende  der Liste ein Element anhängen
<br>
Die Identifikationsnummer bleibt gleich, auch wenn das Objekt verändert wird. Dies ist bei einem mutable Objekt der Fall.

```python
Liste.append(4)
print(Liste)
type(Liste), id(Liste)
```

Schritt 3: an das Ende der Liste ein Element anhängen

```python
Liste.append(5)
print(Liste)
type(Liste), id(Liste)
```

In alle drei Schritten bleibt die Id der Variablen Liste gleich.
