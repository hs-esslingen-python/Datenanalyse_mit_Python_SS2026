### Format Codes 

Die folgende Methoden zur Ausgabe `strftime()` und zum Einlesen `strptime()` verwenden _Format Codes_:

- [Python strftime cheatsheet ](https://strftime.org/)
- [datetime — Basic date and time types¶](https://docs.python.org/3/library/datetime.html)

### Ausgabe mit `strftime()`

string format time

Format: Jahr-Monat-Tag Stunden:Minuten:Sekunde.Mikrosekunden

+ Jahr %Y (vier Stellen)
+ Monat %m
+ Tag %d
+ Stunden %H
+ Minuten %M
+ Sekunden %S
+ Mikrosekunden %f

```python
now.strftime("Gerade war es: %Y-%m-%d %H:%M:%S.%f")
```

Format: Wochentag Stunden:Minuten Tag.Monat.Jahr

+ Wochentag %A

```python
now.strftime("Gerade war es: %A %H:%M %d.%m.%y")
```

Format: Monat als Worten
+ Monat als Wort %B

```python
now.strftime("Gerade war es: %Y-%B-%d %H:%M:%S.%f")
```

### Einlesen Zeitpunkt aus Zeichenketten `strptime()`

**str**ing **p**arse **time**

Zeichenkette mit Zeitpunkt in ein datetime-Objekt wandeln

```python
zeichenkette = '2020-09-18 14:26:43.58070'
datetime_obj = datetime.strptime(zeichenkette,"%Y-%m-%d %H:%M:%S.%f")
datetime_obj
```

und wieder ausgeben:

```python
datetime_obj.strftime("zeichenkette: %Y-%m-%d %H:%M:%S.%f")
```

Beispiel: Lesen einer Zeitangabe aus einer Messdatei:

```python
zeile = 'Start (date/time): 2020-09-18 14:26:06.00000'
Anfang = 'Start (date/time):'
```

```python
string = zeile.replace(Anfang,'').strip()
string
```

```python
datetime.strptime(string,"%Y-%m-%d %H:%M:%S.%f")
```

alles zusammen:

```python
zeilen = ['Start (date/time): 2020-09-18 14:26:06.00000',
          'END   (date/time): 2020-09-18 14:26:43.58070']

Anfang = 'Start (date/time):'

for zeile in zeilen:
    #print(zeile)
    if Anfang in zeile:
        #print(zeile)
        zeichenkette = zeile.replace(Anfang,'').strip()
        datetime_obj = datetime.strptime(zeichenkette,"%Y-%m-%d %H:%M:%S.%f")
        #print(datetime_obj)
        break
    else:
        datetime_obj = None

print(datetime_obj)
```
