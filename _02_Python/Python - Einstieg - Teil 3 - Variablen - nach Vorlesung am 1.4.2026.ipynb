## Variablenzuweisung

Der Datentyp von Variablen wird zur Laufzeit festgelegt und kann auch im Laufe des Programms verändert werden.

Python nutzt eine [dynamische Typisierung](https://de.wikipedia.org/wiki/Dynamische_Typisierung).
Bei der dynamischen Typisierung wird der Datentyp von Variablen zur Laufzeit eines Programms festgelegt.

Im Gegensatz dazu gibt es die [statische Typisierung](https://de.wikipedia.org/wiki/Statische_Typisierung), hier wird der Datentyp von Variablen schon während der Kompilierung festgelegt, z.B. bei den Programmiersprachen C, C++. Statische Typisierung kann durch [starke Typisierung](https://de.wikipedia.org/wiki/Starke_Typisierung) erweitert werden. 



Benennung von Variablen (siehe [www.python-kurs.eu](https://www.python-kurs.eu/python3_variablen.php))

### Aber Funktionen lassen sich überschreiben - Beispiel print-Function

Variablen sind Referenzen auf Objekte.
Objekte liegen im Speicher und speichern Inhalte, wie Datentypen oder auch Funktionen.

Der `print()`-Befehl ist eine eingebaute Funktion, die als Objekt im Speicher liegt. 
<br>
Der Bezeichner `print` zeigt darauf. Mit `type()` bekommen wir einen Hinweis. Das Objekt ist vom Typ _builtin_function_or_method_
<br>
Hier wird `print` ohne Klammern verwendet `()`. 
<br>
Die Klammer `()` werden benötigt, um eine Funktion auszuführen.

````python
type(print)
````

#### Unterschied `print()` und `print`

`print()` -> der print Befehl wird ausgeführt
<br>
`print` -> das ist das Objekt, was den print Befehl ausführen kann


Typ des Rückgabewerte, wenn der `print()`-Befehl ausgeführt wird. 

````python
type(print())
````

Der `print()`-Befehl gibt nichts zurück, deshalt _NoneType_

````python
ret = print('Hello World')
print(ret)
````

`print()` ist eine Funktion die _nichts_ zurückgibt. Dies wird in Python durch `None` zum Ausdruck gebracht.

````python
type(None)
````

#### Variablen verweisen auf Objekete (Referenz)

Wir können eine Variable auf eine andere Variable kopieren. Es wird nicht der Inhalt der Variablen kopiert, sondern nur die Referenz auf dasselbe Objekt.
Für die Zuweisung der Referenz werden keine Klammern `()` verwendet.

````python
tmp_print = print
````

Jetzt test wir, wie `print()` und `tmp_print()` reagieren.

````python
print("Hello world")
````

````python
tmp_print("Hello world mit tmp_print")
````

Jetzt werden wir `print` mit 3 überschreiben

`print()` funktioniert wie vorher und `tmp_print()` funktioniert auch wie `print()`

````python
print=3
print
````

````python
print("Hello world")
````

Jetzt funkioniert `print()` nicht mehr wie gewohnt.
<br>
Aber `tmp_print()` funktioniert noch wie `print()` vorher.

````python
tmp_print("Hello world")
````

Jetzt kopieren wir `tmp_print` auf `print` und dann tut `print()` wieder wie gewohnt.

````python
print = tmp_print
print('Hello World')
````

#### Print ohne Zeilenumburch an Ende

````python
print('Hello ')
print('World')
````

Um bei `print()` den Zeilenumbruch zu unterdrücken, muss ein zusätzlicher Parameter `,end=''` verwendet werden:

````python
print('Hello ',end='')
print('World')
````
