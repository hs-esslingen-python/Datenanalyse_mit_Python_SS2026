# Python - Klassen - Einstieg

[Wikipedia - Klassen](https://de.wikipedia.org/wiki/Klasse_(Objektorientierung))

Unter einer **Klasse** (auch Objekttyp genannt) versteht man in der objektorientierten Programmierung ein abstraktes Modell bzw. einen Bauplan für eine Reihe von ähnlichen **Objekten**.

Die Klasse dient als Bauplan für die _Abbildung von realen Objekten in Softwareobjekte_ und beschreibt 
- **Attribute** (Eigenschaften) und 
- **Methoden** (Verhaltensweisen) der Objekte. 

[Wikipedia - Objekt](https://de.wikipedia.org/wiki/Objekt_(Programmierung))

Ein **Objekt** ist ein konkretes Exemplar (Instanz), das gemäß diesem „Bauplan“ (**Klasse**) erstellt wurde.

Ein Objekt wird während der Laufzeit erzeugt (Instanziierung), d. h. es wird Speicher reserviert, in dem seine genauen Eigenschaften gespeichert werden können. 
Zusammen mit den zugehörigen Methoden ist dieser Speicherbereich das Objekt.

## Erstellen einer Klasse am Beispiel _Messwert_

Ein Messwert soll durch folgende Eigenschaften (Attribute) beschrieben werden:
- Name 
- Wert
- Einheit
- Zeitpunkt

und soll folgende Verhaltensweisen (Methoden) besitzen:
- aktualisieren
- auslesen

Laden des `datetime`-Moduls zur Verwaltung der Zeitpunkte

```python
from datetime import datetime
```

### Definition der Klasse (Bauplan)

Die Klasse wollen wir `Messwert` nennen.

Die Definition einer Klasse wird mit dem Schlüsselwort `class` eingeleitet, gefolgt von dem Namen der Klasse, in unserem speziellen Fall `Messwert`.
Im Gesatz zur Funktion benötigt die Klasse zunächt keine runden Klammern. 
Werden in runden Klammern ein oder mehrere Bezeichner angegeben, so sind dies weitere Klassen, von denen diese Klasse erbt. Die Vererbung ist eine weitere wichtige Eigenschaft von Klassen, die wir hier aber nicht behandeln wollen.
<br>
Innnerhalb der Klassendefinition folgen jetzt eingerückt Definitionen von Funktion, was die Methoden der Klasse sind.
<br>
Es gibt eine spezielle Methode, die [`__init__()`](https://cito.github.io/byte_of_python/read/class-init.html)-Methode, die automatisch aufgerufen wird, wenn ein Objekt einer Klasse angelegt wird und dafür sorgt, dass innerhalb des Objekts die Attribute initialisiert werden.
Methoden innerhalb einer Klasse, die mit __ (zwei Unterstrichen) beginnen und enden, werden als 'dunder'-Methoden (double underscore) bezeichnet und haben spezielle Bedeutungen in Python.

Dann folgen die weiteren Methoden der Klasse, in unseren Fall sind dies die Methoden `aktualisieren()` und `auslesen()`, die für die Kommunikation mit außerhalb des Objekts verwendet werden.

In Python ist `self` eine Konvention für den ersten Parameter einer Methode innerhalb einer Klasse. Es ist eine Referenz auf die aktuelle Instanz der Klasse und wird verwendet, um auf Attribute oder Methoden zuzugreifen, die zu dieser speziellen Instanz der Klasse gehören.

Beim Aufruf einer Methode wird der Parameter nicht angegeben, sondern von Python automatisch ergänzt.


```python
class Messwert:
    
    def __init__(self, Name, Einheit):
        self.Name = Name
        self.Einheit = Einheit
        self.Wert= None
        self.Zeitpunkt = None
    
    def aktualisieren(self, Wert):
        self.Wert = Wert
        self.Zeitpunkt = datetime.now()
        
    def auslesen(self):
        if self.Wert:
            return f"{self.Name}:  {self.Wert} {self.Einheit} von {self.Zeitpunkt}"
        else:
            return f"{self.Name} noch nicht aktualisiert"
```

### Anlegen von Objekten (Instanzen)

Von der Klasse `Messwert` wird ein Objekt `inst_druck` erstellt.
Intern wird die  `__init__(self, Name, Einheit)`-Methode aufgerufen.
<br>
Hinweis: der Parameter `self` taucht in der Argumentliste nicht mehr auf; nur die Argumente `Name` und `Einheit` werden der Methode übergeben:

```python
inst_druck = Messwert('Druck','bar') 
```

```python
inst_druck
```

Von welchem Typ ist die Instanz?
<br>
Die Instanz ist von unserem selber definierten Datentyp `Messwert`:

```python
type(inst_druck)
```

### Methoden

Methode `.auslesen()` aufrufen

```python
inst_druck.auslesen()
```

Methode `.aktualisieren()` mit Parameter für das Argument `Wert` aufrufen:

```python
inst_druck.aktualisieren(5)
```

und wieder die Methode `.auslesen()` aufrufen

```python
inst_druck.auslesen()
```

```python
inst_druck.aktualisieren(7)
```

```python
inst_druck.auslesen()
```

Um alle Methoden und Attribute der Instanz `inst_druck` ausgeben zu lassen, können wir die `dir()`-Funktion verwenden:

```python
dir(inst_druck)
```

Wir legen weitere Instanzen `inst_temperatur` und `inst_geschwindigkeit` an.
<br>
Dies geht jetzt relativ einfach. Dies ist ein großer Vorteil von Klassen gegenüber Funktionen:

```python
inst_temperatur = Messwert('Temperatur','°C') 
inst_geschwindigkeit = Messwert('Geschwindigkeit','km/h') 
```

```python
inst_temperatur.aktualisieren(25.5)
inst_temperatur.auslesen()
```

```python
inst_geschwindigkeit.aktualisieren(50.0)
inst_geschwindigkeit.auslesen()
```

### Attribute

Der Zugriff auf die __Attribute__ des Objekts ist möglich über die Punkt-Notation.

```python
inst_druck.Name
```

```python
inst_druck.Einheit
```

```python
inst_druck.Zeitpunkt
```

Und können sogar von außen geändert werden.
<br>
Einen Schutz bietet Python hier nicht, wie beispielsweise C++, wo public und private Attribute gibt.
<br>
In Python gibt es die Konvention, dass private Attribute, die nur innerhalb der Klassen verwendet werden, aber von außen nicht, mit einem Underscore `_` beginnen.

```python
inst_druck.Name='Druck2'
```

```python
inst_druck.Name
```

```python
inst_druck.auslesen()
```
