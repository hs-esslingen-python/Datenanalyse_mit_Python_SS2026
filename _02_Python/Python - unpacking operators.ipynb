## Anwendungsbeispiele unpacking operators `*` und `**`

- operator `*`  -> Auspacken von Iterables (Listen und Tuples) in Positions-Argumente
- operator `**` -> Auspacken von Dictionries in Keyword-Argumente

#### operator `*`  -> Auspacken von Iterables in Positions-Argumente

Beispiel: Auspacken einer Liste und Ausgabe mit `print()`

```python
my_list = [1, 2, 3]
print(my_list)
print(*my_list)
```

Anwendungsbeispiel: Die einzelnen Argumente bei einem Funktionsaufruf durch eine Liste befüllt werden:

```python
def my_sum(a, b, c):
    """
    This function takes three arguments and returns their sum.
    :param a: The first number.
    :param b: The second number.
    :param c: The third number.
    :return: The sum of the three numbers.
    """
    print(a + b + c)
```

```python
my_list = [1, 2, 3]
```

```python
my_sum(my_list[0],my_list[1],my_list[2])
```

```python
my_sum(*my_list)
```

Anwendungs-Beispiel: Aufspalten einer Liste in mehrere Teile

Fall1: erstes Element der Liste in `a` und letztes Element der List in `c` ablegen. 
Den mittleren Teil in `b` ablegen:

```python
my_list = [1, 2, 3, 4, 5, 6]

a, *b, c = my_list

print(a, type(a))
print(b, type(b))
print(c, type(c))
```

Fall2: ersten Element der Liste in `a`; den Rest der Liste in `b`:

```python
a, *b = my_list

print(a, type(a))
print(b, type(b))
```

Fall3: letztes Element der Liste in `b`, die Teilliste davor in `a`:

```python
*a,b = my_list

print(a, type(a))
print(b, type(b))
```

Anwendungsbeispiel: zwei Listen zusammenfassen

```python
my_first_list = [1, 2, 3]
my_second_list = [4, 5, 6]
my_merged_list = [*my_first_list, *my_second_list]

print(my_merged_list)
```

alternative Zusammenfassung von Listen mit der `extend()`-Methode:

```python
my_first_list.extend(my_second_list)
my_first_list
```

oder mit dem `+` Operator

```python
my_first_list = [1, 2, 3]
my_second_list = [4, 5, 6]
my_first_list + my_second_list
```

Zeichenkette in eine Liste mit Einzelzeichen zerlegen

```python
a = [*"RealPython"]
print(type(a))
print(a)
```

Alternative:

Das Komma hinter `*a` ist entscheidend, sonst kommt es zu einer `SyntaxError`-Fehlermeldung.

```python
*a, = "RealPython"
print(type(a))
print(a)
```

```python
*a = "RealPython"
```

#### operator `**` -> Auspacken von Dictionries in Keyword-Argumente

direktes Anwenden des `**` führt zu einer `SyntaxError`-Fehlermeldung:

```python
my_first_dict = {"A": 1, "B": 2}
**my_first_dict
```

Definieren einer Funktion die keyword Argumente entgegennehmen kann (siehe weiter oben):

```python
def take_kwargs(**kwargs):
    """
    This function takes a variable number of keyword arguments and prints them.
    :param kwargs: A variable number of keyword arguments.
    """  
    print(f"Schlüsselwort-Argumente kwargs: {kwargs}")    
```

```python
my_first_dict = {"A": 1, "B": 2}
take_kwargs(**my_first_dict)
```

Anwendungsbeispiel: zwei Dictionaries zusammenfassen

```python
my_first_dict = {"A": 1, "B": 2}
my_second_dict = {"C": 3, "D": 4}
my_merged_dict = {**my_first_dict, **my_second_dict}

print(my_merged_dict)
```

alternative Zusammenfassung von Dictionary mit der `update()`-Methode:

```python
my_first_dict.update(my_second_dict)
my_first_dict
```
