### Methoden von Zeichenketten 

```python
string = 'Methoden von Objekten #Objekte'
```

`.lower()` alles in Kleinbuchstaben wandeln

`.upper()` alles in Großbuchstaben wandeln

```python
string.lower()  
```

```python
string  # string bleibt unverändert
```

```python
string.upper()  
```

`.split(separator)` Zeichenkette an den Trennzeichen (separator) in eine Liste zerlegen

```python
string.split()  # an den Leerzeichen trennen und eine Liste erzeugen
```

```python
string.split('#') # am Hash(#) trennen und eine Liste erzeugen
```

```python
string.split('#')[1] # zweites Element der Liste
```

Beispiel Email

```python
email= "xyz@gmx.de"
```

```python
liste = email.split("@")
liste
```

```python
liste[1]
```

```python
email.split("@")[1]
```

Entfernen von Leerzeichen (Whitespace)

```python
string2 = ' 12.5  '
```

Die Funktion `float` toleriert Leerzeichen rund um die eigentlichen Fließkommazahl.

```python
float(string2)
```

```python
string2.strip()
```

nur voranstehende Leerzeichen entfernen

```python
string2.lstrip()
```

nur nachgestellte Leerzeichen entfernen

```python
string2.rstrip()
```

```python
string3 = "12.5 bar"
```

Beispiel: Messwert mit Fließkommazahl und Einheit

```python
float(string3)
```

Die Funktion `float` toleriert die Einheit nicht. Aber wir können in einem ersten Schritt die Zahl von der Einheit mit der `.split()`-Methode trennen. 

```python
zahl = float(string3.split()[0])
zahl, type(zahl)
```

```python
zahl, einheit = string3.split()
zahl = float(zahl)
print(f'{zahl=}, {einheit=}')
print(f'{type(zahl)=}, {type(einheit)=}')
```

### `in` - Operator

überprüft ob ein gegebenes Element in einer Liste oder einem Tupel vorkommt 

```python
'x' in [1,2,3]
```

```python
'x' in ['x','y','z']
```

```python
'x' in ('x','y','z')
```

überprüft ob eine gegebene Zeichenkette in einer anderen Zeichenkette vorkommt 

```python
'Test' in 'Orginal'
```

```python
'gina' in 'Orginal'
```
