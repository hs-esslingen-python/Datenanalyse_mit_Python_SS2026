## raw strings

_Raw Strings_ beginnen mit einem `r` vor den Anführungszeichen

Bei raw strings wird die Folge Backlashes+Zeichen `\<Zeichen>` nicht in ein Sonderzeichen umgewandelt.

Bei Standard Strings werden der Backlash `\` gefolgt von einem Zeichen als Sonderzeichen [Escape Characters](https://python-reference.readthedocs.io/en/latest/docs/str/escapes.html) interpretiert: 
- `\n`  newline - neue Zeile
- `\r`  carriage return - Zeilenrücklauf
- `\t`  tabulator
- `\\`  backlash als Zeichen selber



Pfad als Standard-Zeichenkette -> funktioniert nicht wie gewünscht

```python
s = 'C:\new\path\table'
print(s)
```

die `\` müssen verdoppelt werden ("Escape Characters")

```python
s = 'C:\\new\\path\\table'
print(s)
```

Verwendung eines raw strings, um Pfade "natürlich" angeben zu können.

```python
s= r'C:\new\path\table'
print(s)
```

```python
s= r'C:\new\path\table' \
'zweite Zeile ' \
'und noch eine dritte Zeile'
s
```

```python
print(s)
```

## formatierte Ausgabe

mit `str.format()`

```python
zahl=42
string='Anton'
```

Ich möchte folgenden Satz ausgeben:
<br>
_Anton ist 42 Jahre_
aber den Namen und die Zahl aus den Variablen `string` bzw. `zahl` verwenden:

```python
print(string,'ist',zahl,'Jahre')
```

Wo kommen die Leerzeichen zwischen den Element her?

Der `print()`-Befehl besitzt noch einen "versteckten" Parameter `sep` (für separator - Trenner) mit dem dies vorgegeben wird:
Der Parameter `sep` ist standardmäßig ein Leerzeichen.

```python
print(string,'ist',zahl,'Jahre alt',sep='-')
```

```python
print(string,'ist',zahl,'Jahre alt',sep='')
```

formatierte Ausgabe mit `.format()`
<br>
Idee: Baue in die Zeichenkette Platzhalter für die Ausgabe von Variablen ein.
Die Platzhalter werden in Form von Paaren geschweifter Klammern `{}` dargestellt.
Die Variablen werden dann durch die `.format()`-Anweisung übergeben.


```python
print('{} ist {} Jahre'.format(string, zahl))
```

Wiederverwendung der Argumente in der `.format()`-Anweisung durch Indexierung in den geschweiften Klammern z.B. `{0}`  für erstes Argument. 

```python
print('{0} ist {1}, {1} Jahre ist {0}'.format(string, zahl))
```

Neben der Indexierung ist auch die Verwendung von Name möglich. 
<br>
Übergeben ich eine Methode oder eine Funktion Parameter mit Name, dann nennt man dies _keyword parameter_.

```python
print('{} {s} ist {z}, {z} Jahre ist {s}'.format(zahl,z=zahl,s=string))
```

## f-strings  (formatted string literals)

seit Python Version 3.6

Bei **f-strings** wird der Zeichenkette ein `f` oder `F` vorangestellt.
<br>
Innerhalb der Zeichenkette werden geschweifte Kammern-Paare `{}` verwendet, in denen jetzt aber Python Ausdrücke mit Variable stehen dürfen und deren ausgewertetes Ergebnis angezeigt wird.
<br>
Die Darstellung des Ausdrucks kann durch einen Doppelpunkt `:` gefolgt von einer Formatanweisung modifiziert werden.


Referenzen:
- [Python 3's f-Strings: An Improved String Formatting Syntax (Guide)](https://realpython.com/python-f-strings/)

```python
print(F'{string} ist {zahl} Jahre alt, nächstes Jahr wird er {zahl+1}')
```

#### Spezialfall: Wie gebe ich geschweifte Klammern  `{` oder `}` in einem F-String aus?

Die geschweiften Klammern müssen verdoppelt werden: `{{` bzw. `}}`

Referenz: [Stackoverflow - How do I print curly-brace characters in a string while using .format?](https://stackoverflow.com/questions/5466451/how-do-i-print-curly-brace-characters-in-a-string-while-using-format)

```python
print(F'Dies ist eine geschweifte Klammer {{ und die ist noch eine }} und dies ist die Zahl {zahl}')
```

#### Formattierung und Ausrichtung der Ausgabe

Breite für Ausgabe vorgeben

```python
s='Hallo'
print(f"|1234567890|")
print(f"|{s}|")
print(f"|{s:10}|")     # Breite von 10 Zeichen für Ausgabe vorgeben
print(f"|{s:<10}|")    # Ausgabe linksbündig (ist aber default Einstellung)
print(f"|{s:>10}|")    # Ausgabe rechtsbündig
print(f"|{s:^10}|")    # Ausgabe mittig
```

#### Ausgabe einer Fließkommazahl mit drei Nachkommastellen

Die Formattierung `:.3f` bedeutet, dass das Ergebis als Fließkommazahl ausgegeben werden soll mit drei Stellen nach dem Punkt, der als Dezimaltrenner fungiert.

Es wird gerundet.

```python
Fließkommazahl = 3.1415
print(Fließkommazahl) 
print(f"{Fließkommazahl:.3f}")  
```

Ausgabe einer Fließkommazahl mit insgesamt sieben Zeichen mit `:7.2f`.
<br>
Zwei davon sind die Nachkommastellen und ein Zeichen ist der  Dezimaltrenner `.`.
Dann bleiben 4 Zeichen vor dem Dezimaltrenner.
Wird ein Zeichen nicht mit einer Zahl belegt, bleibt es frei.

```python
Fließkommazahl = 123.4567
print(Fließkommazahl) 
print(f"|1234567|")  # zum besseren Abzählen der 7 Zeichen
print(f"|{Fließkommazahl:7.5f}|")  
print(f"|{Fließkommazahl:7.2f}|")
print(f"|1234567890|")
print(f"|{Fließkommazahl:10.2f}|") 
```

Ausgabe einer sehr großen Fließkommazahl

```python
Fließkommazahl2 = float(2**128)
print(f"{Fließkommazahl2}")  
print(f"{Fließkommazahl2:f}")  # Standardschreibweise
print(f"{Fließkommazahl2:g}")  # Exponetenschreibweise
print(f"{Fließkommazahl2:E}")  # Exponetenschreibweise (Großbuchstaben)
```

Interger

## Zeichenketten und Operatoren 

### '+' Operator

Mit dem `+`- Operator können Zeichenkennen verbunden werden. 

```python
'123' + '456' + '789'
```

Dies geht aber auch ohne `+` Operator

```python
'123' '456' '789'
```

`*` Operator

Mit der Anweisung _Zeichenkette_ `*` _Zahl_ wird die Zeichenketten Zahl-mal ausgegeben.

```python
'123'*3
```

Die Zahl kann auch vorangestellt werden:

```python
3*'123'
```

So kann ich einfach Trennlinien einfügen:

```python
80*'-'
```

```python
print(80*'=')
print('Das Ergebnis lautet: ....')
print(80*'=')
```
