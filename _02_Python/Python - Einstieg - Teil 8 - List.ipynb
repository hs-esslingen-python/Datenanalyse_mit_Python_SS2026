#### Ein Element an das Ende einer Liste anfügen

````python
my_list = ['a','b','c']
````

mit dem `+` Operator:

````python
['a','b','c'] + ['d']
````

````python
my_list + ['d']
````

````python
my_list + ['d','e']
````

Bei Verbinden von zwei Listen mit dem `+` Operator wird eine neue Liste erzeugt.
<br>
Die Liste `my_list` selber bleibt unverändert.

````python
my_list
````

Beide Operanden müssen aber Listen sein, sonst gibt es eine `TypeError`-Fehlermeldung: 

````python
my_list + 'd'
````

Anhängen eines Elementes mit der `append()`-Methode:

````python
my_list.append('d')
my_list
````

Durch den Befehl append() wird ein Element an die Liste angehängt. Die Liste wird verändert.

Was sind Methoden?
<br>
Methoden sind Funktionen, die einem Objekt fest zugeordnet sind und die auf die Daten des Objekts zugreifen können.
Wir werden Methoden beim Thema Funktionen und Klassen ausführlicher behandeln. 

Zwei Listen zusammenfügen mit der `append()`-Methode funktioniert nicht:

````python
my_list = ['a','b','c']
my_list2 = ['d','e']
my_list.append(my_list2)
my_list
````

mit `append()` wird eine Liste wird Elememt in eine Liste eingefügt. Dies ist nicht das was wir hier wollen.

mit `extend()` werden die Elemente einer Liste als Elemente in eine Liste eingefügt:

````python
my_list = ['a','b','c']
my_list2 = ['d','e']
my_list.extend(my_list2)
my_list
````

dies entspricht dem Verhalten des `+` Operators:

````python
my_list = ['a','b','c']
my_list2 = ['d','e']
my_list + my_list2
````

Aber es gibt einen Unterschied:
+ a) der `+` Operator erstellt eine neue Liste
+ b) die `extend()` Methode verändert die ursprüngliche Liste

Weitere Methoden von Listen:

+ `.insert()`- Methode:  Element einfügen an beliebiger Position
+ `.remove()`- Methode:  entfernen eines Elements aus der Liste anhand des Elementnamens
+ `.pop()` - Methode:    letzten Elements aus einer Liste entfernen und zurückgeben
+ `.reverse()` - Methode: Reihenfolge der Liste umkehren
+ `.sort()` - Methode:   Sortieren der Liste 
+ `.index()` - Method:   Index zu einem gegeben Element suchen


Beispiel: `.reverse()`-Methode -> Liste wird durch Ausführen der`.reverse()`-Methode verändert:

````python
my_list = ['a', 'b', 'c', 'd', 'e']
my_list.reverse()
my_list
````

````python
my_list.sort()
my_list
````

````python
idx = my_list.index('d')
idx
````

````python
my_list = ['c', 'e', 'd',  'b', 'a']
my_list.sort()
my_list
````

`sorted()`-Funktion  - gibt eine sortierte Liste zurück; ursprüngliche Liste bleibt unverändert:

````python
my_list = ['c', 'e', 'd',  'b', 'a']
print('sorted(my_list):',sorted(my_list))
print('my_list        :',my_list)
````

````python
type(sorted)
````

Zugriff auf ein Element außerhalb des zulässigen Index-Bereiches fürt zu einer `IndexError`-Fehlermeldung:

````python
my_list[5]
````

Verwenden einer Variable als Index

````python
idx = -1
my_list[idx]
````

Index muss aber eine Ganzzahl, also vom Typ _int_, sein, sonst gibt es eine `TypeError`-Fehlermeldung: 

````python
idx = -1.5
my_list[idx]
````

````python
idx = -1.5
print(int(idx))
my_list[int(idx)]
````

Zugriff mit einer Zeichenkette als Index, liefert eine `TypeError`-Fehlermeldung:

````python
s = '1'
my_list[s]
````

Fehlermeldungen:

- Index Fehler `IndexError:` -> Index außerhalb des zulässigen Bereichs
- Type Fehler `TypeError:` -> Index ist nicht vom Typ Integer

#### Zugriff auf eine Teil-Liste - **Slicing**

`my_list[Start_Index:End_Index:Schrittweite]`

Rückgabe als Liste, anfangen bei 'Start_Index' bis **eins** vor 'End_Index' (nicht mehr enthalten)

```
    Liste           [0, 1, 2, 3, 4]
    Index vorwärts   0  1  2  3  4  5
    Index rückwärts -5 -4 -3 -2 -1  
```
   
Die Default-Schrittweite ist 1. Die Schrittweite kann positive und negative Ganzzahlen annehmen

````python
my_list = [0,1,2,3,4]
````

Ausgabe der ersten drei Werte der Liste:

````python
my_list[0:3]
````

````python
my_list[0:3:1]
````

````python
my_list[0:3:2]
````

Die Null am Anfang kann weggelassen werden:

````python
my_list[:3]
````

Ausgabe ab dem zweiten Element bis zum Schluss der Liste:

````python
my_list[1:]
````

alternativ:

````python
my_list[1:5]
````

````python
my_list[1:-1]
````

Ausgabe der letzten beiden Elemente der Liste:

````python
my_list[-2:]
````

````python
my_list[-2:0]
````

````python
my_list[-2:0:-1]
````

````python
my_list[-2::-1]
````

Ausgabe jedes zweiten Elements der Liste 
- Schrittweite wird auf 2 gesetzt und
- Start und Stop-Werte lasse ich weg (sucht sich Python selber)

```
    Liste           [0, 1, 2, 3, 4]
    Index vorwärts   0  1  2  3  4  5
    Index rückwärts -5 -4 -3 -2 -1  
```
   

````python
my_list[0:5:2]
````

````python
my_list[::2]
````

Beispiel: Start bei Index=1 und Stop bei Index -2 (nicht mehr enthalten):
```
    Liste           [0, 1, 2, 3, 4]
    Index vorwärts   0  1  2  3  4  5
    Index rückwärts -5 -4 -3 -2 -1  
```

````python
my_list[1:-2:2]
````

Beispiele mit negative Schrittweite
```
    Liste           [0, 1, 2, 3, 4]
    Index vorwärts   0  1  2  3  4  5
    Index rückwärts -5 -4 -3 -2 -1  
```

Start bei Index=1 und Stop bei Index 4 (nicht mehr enthalten), aber eine negativen Schrittweite -1, liefert eine leere Liste, weil Start zu Stop nicht zur Schrittweite passt

````python
my_list[1:4:-1]
````

Vertauschen von Start nach Stop liefert jetzt eine gefüllte Liste zurück:

````python
my_list[4:1:-1]
````

Liste umkehren -> Schrittweite auf -1

````python
my_list[::-1]
````

````python
my_list
````

Die Anweisung `my_list[::-1]` verändert die Liste selber nicht.

### Ersetzen einer Teilliste durch eine andere Liste

[How does assignment work with list slices? ](https://stackoverflow.com/questions/10623302/how-does-assignment-work-with-list-slices)

Unter Anwendung der Slicing-Notation bekommt man eine Teilliste

````python
my_list = [0,1,2,3,4]
my_list[1:3]
````

Diese Teilliste in der ursprünglichen Liste kann ich durch eine andere Liste ersetzen:

Fall 1: Die Teilliste [1,2] in der Liste [0,1,2,3,4] wird durch [10] ersetzt: [0, 10, 3, 4]

````python
my_list = [0,1,2,3,4]
my_list[1:3] = [10]   # Liste mit nur einem Element
my_list
````

Fall 2: Die Teilliste [1,2] wird durch [10, 11] ersetzt.

````python
my_list = [0,1,2,3,4]
my_list[1:3] = [10,11] # Liste mit zwei Elementen
my_list
````

Fall 3: Die Teilliste [1,2] wird durch [10, 11, 12] ersetzt.

````python
my_list = [0,1,2,3,4]
my_list[1:3] = [10,11,12] # Liste mit nur einem Element
my_list
````

weiteres Beispiel: Die Teilliste [3, 4] wird durch ['vorletzter','letzer'] ersetzt.

````python
my_list = [0,1,2,3,4]
print(my_list[-2:])
my_list[-2:] = ['vorletzter','letzer']
my_list
````

weiteres Beispiel: Die Teilliste [4] wird durch ['Neu2','Neu3'] ersetzt.

````python
my_list = [0,1,2,3,4]
print(my_list[-1:])
my_list[-1:] = ['Neu2','Neu3']
my_list
````

weiteres Beispiel: Die Teilliste [1] wird durch ['Wert1','Wert2'] ersetzt.

````python
my_list = [0,1,2,3,4]
print(my_list[1:2])
my_list[1:2] = ['Wert1','Wert2']
my_list
````
