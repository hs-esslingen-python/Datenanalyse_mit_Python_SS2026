#### Unterscheidung zwischen dem Identity Operator und dem Equality Operator

Der identity operator `is` prüft, ob beide Variablen auf dasselbe Objekt im Speicher zeigen (Identität)
<br>
Der equality operator `==` prüft, ob die Werte zweier Variablen gleich sind (Wertgleichheit).

Fall 1: zwei Variable `x` und `y` verweisen jeweils auf Objekte mit dem gleichen Inhalt, in dem Fall eine Liste mit zwei Elementen.

````python
x=[1,2]
y=[1,2]
````

Der Inhalt ist gleich

````python
x==y
````

Aber es handelt sich um unterschiedliche Objekte

````python
x is y
````

Wir geben die eindeutigen Identität der Objekte `x` und `y` aus:

````python
id(x)
````

````python
id(y)
````

Die unterschiedlichen eindeutigen Identitäten belegen, dass `x` und `y` auf verschiedene Objekte verweisen.

### Logische Operatoren

Es gibt die folgenden logischen Operatoren in Python

+ `and`: Logisches Und – Gibt True zurück, wenn beide Operanden wahr sind, sonst False.

+ `not`: Logische Negation – Gibt das Gegenteil des Wahrheitswerts des Operanden zurück (True wird zu False und umgekehrt)

Die logischen Operatoren liefern einen boolschen Wert (`False`, `True`) zurück.

````python
True and True
````

````python
True and False
````

````python
(1 > 2) or (2 < 3)
````

````python
(1 == 2) or (2 == 3) or (4 == 4)
````

````python
not (1 > 2) 
````

````python
not((1 > 2) and (2 < 3))
````

### Bitweise Operatoren 

In Python gibt es sechs bitweise Operatoren, die direkt auf der Bit-Ebene von Ganzzahlen arbeiten. 

+ `&` Bitweises UND  - Setzt jedes Bit auf 1, wenn beide Bits 1 sind
+ `|` Bitweises ODER - Setzt jedes Bit auf 1, wenn mindestens eines der Bits 1 ist
+ `^` Bitweises XOR - Setzt jedes Bit auf 1, wenn die Bits unterschiedlich sind
+ `~` Bitweises Komplement (NOT) - Kehrt alle Bits um 
+ `<<` Linksverschiebung - Verschiebt die Bits nach links (entspricht Multiplikation mit 2^n)
+ `>>` Rechtsverschiebung - Verschiebt die Bits nach rechts (entspricht Division durch 2^n)

zunächst arbeiten wir mit boolschen Werten; dann arbeiten die bitweisen Operatore wie die logischen

````python
True & True
````

````python
(1 < 2) & (2 < 3)
````

````python
(1 > 2) | (2 < 3)
````

````python
True ^ True
````

````python
n = 2
a = 0xF
print(f"a = {a:3d}d = {a:09b}b = {a:04x}x")
c = a^(1<<n)
print(f"c = {c:3d}d = {c:09b}b = {c:04x}x")
c = c^(1<<n)
print(f"c = {c:3d}d = {c:09b}b = {c:04x}x")
````

## for-Schleifen

For-Schleifen werden verwendet für
- Iterationen - Es liegt eine Sequenz von Objekten vor (z.B. eine Liste), die ich nacheinander auf dieselbe Art und Weise bearbeiten möchte
- Wiederholungen (engl. repetitions) - Ich möchte etwas mehrfach ausführen

Im folgenden werden Iterationen behandelt.
Die Wiederholungen (engl. repetitions) werden in Zusammenhang mit der Vorstellung von `range()` vorgestellt.


#### Iteration, Iterator und Iterables

+ **Iteration** (Wiederholung)
Eine Iteration beschreibt einen Prozess mehrfachen Wiederholens gleicher oder ähnlicher Handlungen
+ **Iterator**
Ein Iterator ist ein Objekt mit einer zählbaren Menge an Objekten, die nacheinander ausgegeben werden.
+ **Iterable**
Ein Iterables ist ein Objekt, welches aus einer Sammlung von Elementen besteht, die nacheinander verarbeitet werden können.

Folgende Datentypen in Python sind Iterables:
+ Liste (list)
+ Tupel (tuple)
+ Menge (set)
+ Zeichenkette (str)
+ Dictionary, assoziative Liste  (dict)


Wir nehme eine Liste mit 3 Elementen. Dies ist ein **Iterable** (iterierbares Objekt).

````python
Iterable = [1,2,3]
Iterable
````

Die in Python fest eingebaute Funktion [`iter()`](https://docs.python.org/3/library/functions.html#iter) erstellt aus einem **Iterable** (iterierbares Objekt) einen **Iterator** (Objekt).

````python
Iterator = iter(Iterable) 
Iterator
````

Mit der fest in Python eingebauten Funktion [`next()`](https://docs.python.org/3/library/functions.html#next), kann ich das "nächste" Element aus dem Iterator herausholen.

````python
next(Iterator)
````

Der Aufruf `next(Iterable)` kann solange wiederholen werden, bis der Iterator keine Elemente mehr hat. Dann gibt Python eine Fehlermeldung `StopIteration`.
Es findet also eine **Iteration** (Wiederholung) statt.

````python
next(Iterator)
````

````python
next(Iterator)
````

````python
next(Iterator)
````

````python
try:
    next(Iterator)
except StopIteration:
    print("Iterator ist am Ende")
else:
    print("Iterator ist nicht am Ende")
````

Eine for-Schleife fasst die Erstellung des Iterators aus einem Iterable (iterierbares Objekt) mit der Funktion `iter()` und das wiederholte Aufruf von `next()` bis der Iterator leer ist zusammen.

#### Definition einer for-Schleife:
```Python
for Element in Iterable:
    <Schleifenrumpf> verwendet Interations-Variablen "Element"
```
Das Keyword `for` leitet die Definition der Schleife ein, gefolgt von der "Interations"-Variablen, hier _Element_, gefolgt vom Keyword `in` und dem iterierbaren Objekt (Iterable), der Sammlung von Elementen, über die iteriert werden soll. Die Zeile wird mit einem Semicolon `:` abgeschlossen.
Danach folgt eingerückt der Rumpf der for-Schleife.
Der Schleifenrumpf der for-Schleife enthält die Anweisungen, die für jede Iteration ausführt werden soll. 
Der Schleifenrumpf wird durch Einrücken von 4 Leerzeichen gekennzeichnet.
In der Programmiersprache C werden zum Markieren des Funktionsrumpfs geschweifte Klammern verwendet; diese gibt es so in Python nicht, stattdessen werden Blöcke von Anweisung durch Einrücken gekennzeichnet. Geschweifte Klammer werden aber bei Sets und Dictionaries verwendet.


````python
Liste = [1,2,3]
for Element in Liste:
    print(Element)
````

Anstelle einer Liste kann auch eine Zeichenkette als Iterable verwendet werden:

````python
Zeichenkette = "Hallo"
for Element in Zeichenkette:
    print(Element)
````

Zusätzlich soll die Nummer des aktuellen Schleifendurchlaufs ausgegeben werden:
<br>
Hierzu wird vor der Schleife eine Variable für einen Zähler angegen und bei jeder Iteration der Schleife, um eins erhöht.

````python
Liste = ['a','b','c']
k=1
for Element in Liste:
    print(k,Element)
    k = k+1
````

### Schleifenzähler mit `enumerate()`

Die [enumerate](https://docs.python.org/3/library/functions.html#enumerate)-Funktion ist eine integrierte Funktion, die in Python fest eingebaut ist. 
Die Funktion [enumerate](https://docs.python.org/3/library/functions.html#enumerate) nimmt ein iterierbares Objekt entgegen und fügt nacheinander jedem Element des iterierbaren Objekts einen Zähler hinzu und gibt das Paar als ein Tupel zurück.


````python
enumerate(Liste)
````

Python erzeugt die Sequenz zunächst nicht, sondern wartet ab, was wir weiter damit machen wollen.
<br>
Dies nennt sich [Lazy Evaluation](https://de.wikipedia.org/wiki/Lazy_Evaluation)
<br>
Erst wenn wir eine Liste daraus erstellen wollen, erzeugt Python diese Sequenz an Tupeln.

````python
list(enumerate(Liste))
````

Jetzt können wir über diese Sequence (Iterable) iterieren:

````python
for Element in enumerate(Liste):
    print(Element)
````

Aber es ist immer noch unpraktisch, weil in der Variable _Element_ das Tupel enthält.
Wie wir die direkt auf zwei Variablen aufteilen können erfahren wir im nächsten Abschnitt _tuple unpacking_.

### tuple unpacking

Zuweisen der Elemente eines Tupels auf einzelne Variablen: 

````python
(k,Element) = (0, 'a')
print(k)
print(Element)
````

Bei der Zuweisung geht es auch ohne Klammern

````python
k,Element = 0, 'a'
print(k)
print(Element)
````

Anstatt einer Iterations-Variablen verwenden wir jetzt zwei Iterations-Variablen und entpacken damit das Tupel.

````python
for k,Element in enumerate(Liste):
    print(f"{k}.Element ist {Element}")
````

Es ist noch unschön, dass der Zähler bei Null und nicht bei Eins startet.
Dies kann aber durch den keyword Parameter `start=1` geändert werden.

````python
for k,Element in enumerate(Liste, start=1):
    print(f"{k}.Element ist {Element}")
````

Beispiel: Verwendung von Positionargumenten

````python
for k,Element in enumerate(Liste, 1):
    print(f"{k}.Element ist {Element}")
````

Beispiel: Verwendung von Keywordargumenten

````python
for k,Element in enumerate(start=1, iterable = Liste):
    print(f"{k}.Element ist {Element}")
````

#### Schlüsselworte **break** und **continue**
`break` - sofortiger Abbruch der Schleife

`continue` - sofort mit der nächsten Iteration beginnen

Beispiel mit `break`:

````python
Liste = [1,2,3,4,5]
for Element in Liste:
    print(f"Element: {Element}")
    if Element==3:
        print(f"3 gefunden -> Element: {Element} -> Abbruch")
        break
print('nach der For-Schleife')
````

Beispiel mit `continue`:

````python
Liste = [1,2,3,4,5]
for Element in Liste:
    print(f"Element: {Element} vor Abfrage")
    if Element==3:
        print(f"3 gefunden -> Element: {Element} -> direkt zu nächsten Iteration")
        continue
    print(f"Element: {Element} nach Abfrage")
print('nach der For-Schleife')
````

Hinweis `Element: 3 nach Abfrage` wird nicht ausgeführt

## while Schleife

Definition einer While-Schleife:

```Python
while <Bedingung>:
    <Schleifenrumpf>
```
Das Keyword `while` leitet die While-Schleife ein, gefolgt von einer Bedingung, die erfüllt sein muss, damit der Schleifenrumpf ausgeführt wird.
Die Zeile wird mit einem Semikolon `:` abgeschlossen.

Beenden der Schleife: Im Schleifenrumpf muss die Bedingung sich verändern oder es muss aus der Schleife gesprungen werden.

Bei der While-Schleife ist vorher nicht bekannt, wie oft sie durchlaufen wird.

Beispiel einer While-Schleife, deren Bedingung beim Durchlaufen verändert.

````python
i = 1
while i < 5:
    #print('i is: {}'.format(i))
    print(f'i is: {i}')
    i = i+1
print('nach der While-Schleife')
````

Beispiel einer While-Schleife, die eine Endlosschleife ist und aus der mittels `break` herausgesprungen wird.

````python
i = 1
while True:
    print(f'1.Pos i is: {i}')
    if i>5:
        print("Bedingung erfüllt")
        break
    else:
         print("Bedingung nicht erfüllt")
    #print('i is: {}'.format(i))
    print(f'2.Pos i is: {i}')
    i = i+1
print('nach der While-Schleife')
````
