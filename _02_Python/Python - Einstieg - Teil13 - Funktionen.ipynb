## Funktion mit Namen - standard function defintion

[Wikipedia - Funktionen](https://de.wikipedia.org/wiki/Funktion_(Programmierung))

Mit einer Funktion (englisch: function) wird der Programm-Quellcode strukturiert, so dass sich Teile der Funktionalität des Programms kapseln und wiederverwenden lassen. 

Eingänge der Funktion werden als Parameter in der Argumentliste übergeben.

Die Funktion gibt direkt ein Resultat zurückgibt und kann deshalb in Ausdrücken verwendet werden. 

In Python können die Argumente einen Defaultwert besitzen und müssen dann nicht beim Funktionsaufruf übergeben werden.

Bei Funktionsaufruf unterscheidet Python zwischen 
- "Positions-Argumenten" (positional arguments) und 
- "Schlüsselwort-Argumenten" (keyword arguments).

Bei "Positions-Argumenten" (positional arguments) muss die korrekte Reihenfolge der Argumente beim Funktionsaufruf eingehalten werden.

Hingegen nennen "Schlüsselwort-Argumenten" (keyword arguments) die Argumente explizit beim Namen und können dann in einer beliebigen Reihenfolge beim Funktionsaufruf übergeben werden.

Beim Funktionsaufruf müssen zuerst die "Positions-Argumente" (positional arguments) gefolgt von den "Schlüsselwort-Argumenten" (keyword arguments) übergeben werden.

### Definition einer Funktion

Eine Funktion wird in Python wie folgt definiert:
```
def <Funktionsname> (Parameterliste):
    """
    Funktionsbeschreibung -> "Docstring"
    """
    <Python Anweisungen - "Rumpf">
    return <Wert>
```
Das Schlüsselwort `def` startet die Funktionsdefinition, gefolgt vom Namen der Funktion und von der Aufzählung der Parameter der Funktion in Klammern, getrennt jeweils durch ein Komma. Die Zeile wird mit einem Doppelpunkt `:` abgeschlossen.
<br>
Für den Funktionsnamen gelten die gleichen Regeln wie für Variablen.
<br>
In den nächsten Zeilen folgt der Funktionsrumpf, der die Anweisungen enthält, die die Funktion ausführen soll. 
Der Funktionsrumpf wird durch Einrücken von 4 Leerzeichen gekennzeichnet.
In der Programmiersprache C werden zum Markieren des Funktionsrumpfs geschweifte Klammern verwendet; diese gibt es in Python nicht, stattdessen werden Blöcke von Anweisung durch Einrücken gekennzeichnet.
<br>
Eingeleitet mit dem Schlüsselwort `return` wird der Rückgabewert der Funktion zurückgegeben. 
<br>
Nach der Ausführung von `return` wird die Funktion beendet, d.h. nachfolgende Anweisung werden nicht mehr ausgeführt, auch wenn sie da stehen.
<br> 
Die Rückgabe mit `return` ist optional.
Anwendungsfälle könnten sein, dass man in einer Funktion nur was mit `print()` ausgeben oder in eine Log-Datei schreiben möchte.
Ein Funktion ohne expiliziten Rückgabewert gibt aber `None` zurück. 
<br>
Weitere Details zu `None`: [Null in Python: Understanding Python's NoneType Object](https://realpython.com/null-in-python/)
<br>


### Funktions-Beispiel:
Im folgenden wird eine Funktion mit dem Name `sub_fkt` und den Argumenten `parameter1` und `parameter2` definiert.
Diese Funktion soll die Differenz aus `parameter1` und `parameter2` zurückgeben.

Definition der Funktion:

````python
def sub_fkt(parameter1, parameter2):
    """
    Subtracts two numbers.
    :param parameter1: The first number to subtract from.
    :param parameter2: The second number to subtract.
    :return: The result of the subtraction.
    """    
    return parameter1 - parameter2
````

Der Funktionsname `sub_fkt` ohne die runden Klammer `()` bezeichnet das Funktions-Objekt:

````python
sub_fkt
````

Mit der eingebauten Python-Funktion `type()` kann der Typ der eben definierten Funktion `sub_fkt` überprüft werden:

````python
type(sub_fkt)
````

Ausgabe des "Docstrings" (Beschreibung der Funktion)

````python
help(sub_fkt)
````

Zum Aufruf der Funktionen werden die runden Klammer `()` verwendet.

### Aufruf der Funktion mit Positionsargumenten (**positional arguments**)

Hier muss die Reihefolge eingehalten werden: zuerst wird der Wert für den `parameter1` und danach der Wert für den `parameter2` übergeben:

````python
sub_fkt(5,2)
````

Werden zu wenig Parameter übergeben, gibt es eine `TypeError`-Fehlermeldung:

````python
sub_fkt(5)
````

### Aufruf der Funktion mit Schlüsselwort-Argumente (**keyword arguments**)

Beim Funktionsaufruf wird den Argumenten explizit durch Zuweisung mit einem Gleichheitszeichen `=` ein Wert zugewiesen:

````python
sub_fkt(parameter1=5, parameter2=2)
````

Der Reihenfolge dieser Zuweisungen ist jetzt beliebig:

````python
sub_fkt(parameter2=2, parameter1=5)
````

### Defaultwerte für Argumente

Bei der Funktionsdefinition können den Argumenten Werte zugewiesen werden. Dann brauchen diese Argumente beim Funktionsaufruf nicht angegeben zu werden.
<br>
**Beachte:** 
Sobald einem Parameter ein Defaultwert zugewiesen wird, müssen den darauf folgenden Parametern auch Defaultwerte zugewiesen werden.

Vorsicht bei Einsatz von mutable-Datentypen wie Listen als Defaultwerte.  

Beispiel: `parameter2` bekommt den Defaultwert 2:

````python
def sub_fkt2(parameter1, parameter2=2):
    return parameter1 - parameter2
````

````python
sub_fkt2(parameter1=5, parameter2=2)
````

````python
sub_fkt2(parameter1 = 5)
````

das funktioniert auch für Positionsargumente (positional arguments)

````python
sub_fkt2(5, 2)
````

````python
sub_fkt2(5)
````

### Argumentliste mit einer beliebigen Anzahl von Argumenten

Referenz: [realpython.com - Python args and kwargs: Demystified](https://realpython.com/python-kwargs-and-args/)

 

#### Beliebig viele Positions-Argumente - positional arguments `*arg`

Durch Voranstellen des __unpacking operator__ (*) werden die beim Funktionsaufruf übergebenen Werte in einem Tuple zusammengefasst:

````python
def my_fkt3(*args):
    """
    This function takes a variable number of arguments and returns their sum.   
    :param args: A variable number of arguments.
    :return: The sum of the arguments.
    """
    print(f"Positions-Argumente args: {args}")
    print(f"Datentyp von args       : {type(args)}")
    
    ergebnis=0  # dies ist eine lokale Variable, die nur innerhalb der Funktion sichtbar ist
    for par in args:
        #ergebnis = ergebnis + par
        ergebnis += par  # ergebnis = ergebnis + par         
        
    return ergebnis
````

Was macht `+=`?

In Python nennt sich der Operator `+=` ein "erweiterter Zuweisungsoperator" oder auch "Verbundzuweisungsoperator". Er wird verwendet, um einer Variablen einen Wert hinzuzufügen und das Ergebnis direkt der gleichen Variablen zuzuweisen. 

````python
k = 0
k +=1
k +=2
k
````

Funktionsaufruf mit 4 Werten:

````python
my_fkt3(1,2,3,4)
````

Funktionsaufruf mit 5 Werten:

````python
my_fkt3(1,2,9,7,8)
````

#### Beliebig viele Schlüsselwort-Argumente - keyword arguments `**kwargs`

Durch Voranstellen des __unpacking operators__ (**) werden die beim Funktionsaufruf übergebenen Werte in einem Dictionary zusammengefasst.

`kwargs` steht für **k**ey**w**ord **arg**ument**s**

````python
def my_fkt4(**kwargs):
    """
    This function takes a variable number of keyword arguments and returns their sum.
    :param kwargs: A variable number of keyword arguments.
    :return: The sum of the keyword arguments.
    """
    print(f"Schlüsselwort-Argumente kwargs: {kwargs}")
    print(f"Datentyp von kwargs           : {type(kwargs)}")
    
    ergebnis=0  # dies ist eine lokale Variable, die nur innerhalb der Funktion sichtbar ist
    for par in kwargs.values():
        ergebnis += par
        
    return ergebnis
````

Funktionsaufruf mit 4 Keyword-Argumenten:

````python
my_fkt4(a=1,b=2,c=3,d=4)
````

Funktionsaufruf mit 5 Keyword-Argumenten:

````python
my_fkt4(a=1,b=2,c=3,d=4,par5=5)
````

Beim Versuch des Funktionsaufrufs mit Positionsargumenten wird eine `TypeError`-Fehlermeldung ausgegeben: 

````python
my_fkt4(1,2,3,4)
````

#### Kombinieren von erforderlichen, optionalen Positions- und Keyword-Argumenten

In der Parameterliste der Funktion gilt die folgende Reihenfolge :
- zuerst die erforderlichen Argumente, danach dann eine 
- beliebige Anzahl von Positions-Argumenten `*args` gefolgt von einer 
- beliebigen Anzahl von Keyword-Argumenten `**kwargs`.

Beispiel:
Die Funktion `my_fkt5` besitzt zwei erforderliche Argumente `a` und `b` und kann weitere optionale Positions- und Schlüsselwort-Argumente entgegen nehmen:

````python
def my_fkt5(a,b,*args,**kwargs):
    """
    This function takes two required arguments, a variable number of positional arguments,
    and a variable number of keyword arguments.
    :param a: The first required argument.
    :param b: The second required argument.
    :param args: A variable number of positional arguments.     
    :param kwargs: A variable number of keyword arguments.
    """
    print(f"erforderliches Argument a     : {a} (Typ {type(a)})")
    print(f"erforderliches Argument b     : {b} (Typ {type(b)})")
    print(f"Positions-Argumente args      : {args} (Typ: {type(args)})")
    print(f"Schlüsselwort-Argumente kwargs: {kwargs} (Typ: {type(kwargs)})")
    
````

Aufruf mit 5 Positions `1,2,3,4,5` und 2 Schlüsselwort-Argumenten `c=6,d=7`:

````python
my_fkt5(1,2,3,4,5,c=6,d=7)
````

Aufruf nur mit den erforderlichen ersten beiden Argumenten `1,2`:

````python
my_fkt5(1,2)
````

Aufruf mit den erforderlichen ersten beiden Argumenten `1,2` und zwei weiteren optionale Positionsargumenten `3,4`:

````python
my_fkt5(1,2,3,4)
````

Aufruf nur mit den erforderlichen ersten beiden Argumenten  `1,2` und zwei weiteren optionalen Keywordargumenten  `c=6,d=7`:

````python
my_fkt5(1,2,c=6,d=7)
````

Aufruf nur mit den erforderlichen ersten beiden Argumenten als Keywordargumenten und zwei weiteren optionalen Keywordargumenten:

````python
my_fkt5(a=1,b=2,c=6,d=7)
````

erforderliche Parameter a und b können als Keywordargument an beliebiger Stelle stehen.

````python
my_fkt5(c=6,d=7,a=1,b=2)
````

Der Versuch mit einem optionalen "keyword argument" zu beginnen führt zu einem `SyntaxError`:

````python
my_fkt5(x=1,2,3,4,5,c=6,d=7)
````

### Lokale und globale Variable

Lokale Varibale werden innerhalb der Funktion zugewiesen; globale Variablen außerhalb der Funktion.

"The keyword `global` is only useful to change or create global variables in a local context, although creating global variables is seldom considered a good solution" aus 
[Why isn't the 'global' keyword needed to access a global variable?](https://stackoverflow.com/questions/4693120/why-isnt-the-global-keyword-needed-to-access-a-global-variable)


Beispiel:
- `x` ist eine globale Variable, auf die auch innerhalb der Funktion zugegriffen werden kann, weil sie innerhalb der Funktion nicht modifiziert wird.
- `y` existiert einmal als globale und einmal als lokale Variable
- `z` ist eine globale Variable, die auch innerhalb der Funktion modifiziert werden darf, weil sie innerhalb der Funktion mit `global z` bekannt gemacht wurde. 

````python
x=1  # globale Variable x
y=2  # globale Variable y
z=3  # globale Variable z

def my_fkt6():
    """
    This function demonstrates the use of global and local variables.
    """
    global z     # die globale Variable z wird innerhalb der Funktion bekannt gemacht 
    
    print(f" 4. innerhalb:  x = {x}")
    print(f" 5. innerhalb:  y = nicht verfügbar")
    print(f" 6. innerhalb:  z = {z}")
    
    y = 20    # erstellen einer lokalen Variablen y
    z = 30    # Modifizieren der globalen Variablen z
    
    print(f" 7. innerhalb:  x = {x}")
    print(f" 8. innerhalb:  y = {y}")
    print(f" 9. innerhalb:  z = {z}")
    
# Aufrufe
print(f" 1. außerhalb:  x = {x}")
print(f" 2. außerhalb:  y = {y}")
print(f" 3. außerhalb:  z = {z}")
my_fkt6()
print(f"10. außerhalb:  x = {x}")
print(f"11. außerhalb:  y = {y}")
print(f"12. außerhalb:  z = {z}")
````

## map and filter

````python
seq = [1,2,3,4,5]
````

#### map
`map(Funktion, Liste)`: Die gegebene Funktion wird auf jedes Element der Liste anwenden und aus den Ergebnissen, wenn gefordert, eine neue Liste zu bauen.

````python
map(times2,seq)
````

auf Wartestellung wird erst ausgeführt, wenn eine Liste angefordert wird

````python
list(map(times2,seq))
````

jetzt das ganze mit der anonymen Funktion durchführen

````python
list(map(lambda var: var*2,seq))
````

#### filter
`filter(Funktion,Liste)`: Die Funktion bekommt nacheinander die Element der Liste übergeben und gibt True oder False zurück. Bei True wird das Element in die neue Liste übernommen, bei False wird es weggelassen.

````python
filter(lambda item: item%2 == 0,seq)
````

`filter()` ist lazy; wir müssen die Auswertung explizit anfordern, in dem wir eine Liste generieren lassen.

````python
list(filter(lambda item: item%2 == 0,seq))
````

Das geht natürlich auch mit einer Standard Python Funktion; jetzt brauche ich aber mehrer Zeilen Python dafür.

````python
def ist_gerade(item):
    return item%2==0
````

````python
list(filter(ist_gerade,seq))
````
